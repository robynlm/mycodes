```python
import h5py
import sys
import time
sys.path.append("/users/munozr/mycodes")
import numpy as np
from pylab import *
import matplotlib as mpl
import matplotlib.pyplot as plt
from data_analysis_codes.tools import ReadingTools as RRead
from data_analysis_codes.tools import GetVars_Plot2d as GVar
from data_analysis_codes.tools import LinData
from matplotlib.collections import LineCollection
from data_analysis_codes.tools import Cstyle
plt.style.use(Cstyle.style1)
```

```python
param32 = RRead.read_parameters('pflrw_d3e2_L1206_t1_N32_EdS')
param64 = RRead.read_parameters('pflrw_d3e2_L1206_t1_N64_EdS')
param128 = RRead.read_parameters('pflrw_d3e2_L1206_t1_N128_EdS')
Lin32 = LinData.LinData_Class(param32)
all_h5it32 = RRead.collect_h5iteration(Lin32.param)
```

```python
def progressbar(it, prefix="", size=60, out=sys.stdout): # Python3.6+
    count = len(it)
    start = time.time() # time estimate start
    def show(j):
        x = int(size*j/count)
        # time estimate calculation and string
        remaining = ((time.time() - start) / j) * (count - j)        
        mins, sec = divmod(remaining, 60) # limited to minutes
        time_str = f"{int(mins):02}:{sec:03.1f}"
        print(f"{prefix}[{u'█'*x}{('.'*(size-x))}] {j}/{count} Est wait {time_str}", end='\r', file=out, flush=True)
    show(0.1) # avoid div/0 
    for i, item in enumerate(it):
        yield item
        show(i+1)
    print("\n", flush=True, file=out)
```

```python
theta32 = []
theta64 = []
theta128 = []

toolbar_width = len(all_h5it32)

for i in progressbar(range(toolbar_width), "Computing: ", toolbar_width):
    it32 = all_h5it32[i]
    
    f = h5py.File('{}_it_{:06d}.hdf5'.format(param32['h5datapath']+param32['simname'], it32), 'r')
    theta32 += [np.array(f['Theta'])]
    f.close()
    
    it64 = it32*2
    f = h5py.File('{}_it_{:06d}.hdf5'.format(param64['h5datapath']+param64['simname'], it64), 'r')
    theta64 += [np.array(f['Theta'])[0::2, 0::2, 0::2]]
    f.close()
    
    it128 = it32*4
    f = h5py.File('{}_it_{:06d}.hdf5'.format(param128['h5datapath']+param128['simname'], it128), 'r')
    theta128 += [np.array(f['Theta'])[0::4, 0::4, 0::4]]
    f.close()
    
theta32 = np.array(theta32)
theta64 = np.array(theta64)
theta128 = np.array(theta128)
```

```python
convergence = abs(theta32 - theta64)/abs(theta64 - theta128)
order = abs(np.log(convergence)/np.log(2))
order[order == np.inf] = np.nan
order[order == -np.inf] = np.nan
order = np.reshape(order, (32*32*32, len(order)))

fig, ax = plt.subplots()
            
emin = np.nanmin(order, axis=0)
e1q = np.nanpercentile(order, 25, axis=0)
e2q = np.nanpercentile(order, 50, axis=0)
e3q = np.nanpercentile(order, 75, axis=0)
emax = np.nanmax(order, axis=0)

plt.plot(all_h5it32, emin, color=cm.get_cmap('Blues')(0.8), linestyle = '-', label='min = {:.2f}'.format(np.min(emin)))
plt.plot(all_h5it32, e1q, color=cm.get_cmap('Blues')(0.65), linestyle = '-', label=r'1st quartile $\simeq$ {:.2f}'.format(np.average(e1q)))
plt.plot(all_h5it32, e2q, color=cm.get_cmap('Blues')(0.5), linestyle = '--', label=r'median $\simeq$ {:.2f}'.format(np.average(e2q)))
plt.plot(all_h5it32, e3q, color=cm.get_cmap('Blues')(0.35), linestyle = ':', label=r'3rd quartile $\simeq$ {:.2f}'.format(np.average(e3q)))
plt.plot(all_h5it32, emax, color=cm.get_cmap('Blues')(0.2), linestyle = '-', label='max = {:.2f}'.format(np.max(emax)))

all_h5it32 = np.array(all_h5it32)
segs = [np.column_stack([all_h5it32, y]) for y in order]
line_segments = LineCollection(segs, array=all_h5it32, cmap = cm.get_cmap('RdPu', 1), alpha=0.01)
ax.set_xlim(np.min(all_h5it32), np.max(all_h5it32))
ax.set_ylim(np.min(emin)-0.5, np.max(emax)+0.5)
ax.add_collection(line_segments)

plt.legend(bbox_to_anchor=(1,1))
plt.grid()
plt.ylabel(r'Convergence order of $\Theta$')
plt.xlabel('Iteration')
plt.show()
```
