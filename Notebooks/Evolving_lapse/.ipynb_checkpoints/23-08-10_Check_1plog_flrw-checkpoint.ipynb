```python
def RK4_tau(t, alpha):
    dt = t[1] - t[0]
    tau = [t[0]]
    alpha_half = NumMethods.Lagrange_interp(t, alpha, t[:-1] + (dt / 2), 
                                            reduce_size=True)
    for i in range(len(alpha)-1):
        k1 = alpha[i]
        k2 = alpha_half[i]
        #k2 = (alpha[i+1]+alpha[i])/2
        k3 = k2
        k4 = alpha[i+1]
        tau += [tau[-1] + (k1 + 2*k2 + 2*k3 + k4) * (dt / 6)]
    return np.array(tau)
```
