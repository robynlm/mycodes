```python
plt.figure(figsize=(17, 5))
loc = 'OD'
anA = Lin.an_initial(fA['t'])
alphafac = 0.2

def plot_old(varname):
    plt.semilogy(anA, abs(fA[varname+loc]))

def plot_new(f, varname):
    varstr = varname+loc
    plt.semilogy(f['an_'+loc], abs(f[varstr]))

ax1 = plt.subplot(141)
plot_old('rho_')
plot_new(fdata, 'rho_u_')
plot_new(fdata_pfix, 'rho_u_')
plot_new(fdata_GRH, 'rho_u_')
plt.grid()
plt.title(r'$|\rho|$'+' at '+loc)
ax1.set(xticklabels=[])

ax2 = plt.subplot(142)
plot_old('K_')
plot_new(fdata, 'K_')
#plot_new(fdata_pfix, 'K_')
#plot_new(fdata_GRH, 'K_')
plot_new(fdata_pfix, 'Theta_')
plot_new(fdata_GRH, 'Theta_')
plt.grid()
plt.title(r'$|\Theta|$')
ax2.set(xticklabels=[])

ax3 = plt.subplot(143)
plot_old('gdet_')
plot_new(fdata, 'gdet_')
plot_new(fdata_pfix, 'gdet_')
#plot_new(fdata_GRH, 'gdet_')
#plot_new(fdata_pfix, 'hdet_')
plot_new(fdata_GRH, 'gdet_')
#plot_new(fdata_GRH, 'hdet3_')
#plot_new(fdata_GRH, 'hdet4_')
plt.grid()
plt.title(r'$\gamma$')
ax3.set(xticklabels=[])

ax4 = plt.subplot(144)
plot_old('RicciS_')
plot_new(fdata, 'RicciS_')
plot_new(fdata_pfix, 'RicciS_')
plot_new(fdata_GRH, 'RicciS_')
plt.grid()
plt.title(r'$|{}^{(3)}R|$')
ax4.set(xticklabels=[])

plt.subplots_adjust(wspace = 0.3)
```
