```python
def get_Amp1(tIN, Amp, Lc, R):
    N = 32
    params = {'Amp_pertx': Amp, 'Amp_perty': Amp, 'Amp_pertz': Amp, 
          'lambda_pertx':Lc, 'lambda_perty':Lc, 'lambda_pertz':Lc}
    output = A.f_Amp(tIN, [Lc]*3, N, params, inhom='sin', expansion='EdS')
    return output['deltaOD'], output['delta1OD']
```

```python
# Previously
R = eds.a_today*(eds.t_today**(-2/3))*1206*eds.Hprop(1)
print('R = ', R)
Amp = [0.01068203445570103244]
L = [1206]

delta = np.zeros((len(Amp), len(L)))
nonlin = np.zeros((len(Amp), len(L)))
NBRDT = np.zeros((len(Amp), len(L)))
for ia, amp in enumerate(Amp):
    for il, l in enumerate(L):
        delt, delt1 = get_Amp1(1, amp, l, R)
        delta[ia, il] = delt
        nonlin[ia, il] = abs(delt / delt1 - 1)
        aF = eds.a(tIN)*1.68/delt
        tF = eds.t_func_a(aF)
        if tF > tIN and eds.z(tF)>0:
            simtime = tF - tIN
            courant_fac = 0.1
            N = 128
            dx = l / N
            dt = courant_fac*np.sqrt(3)*eds.a(tIN)*dx
            NBRDT[ia, il] = simtime / dt
        else:
            NBRDT[ia, il] = -1
print(delt, NBRDT[0,0])
```

```python
R = 10
Amp = [1e-3, 2.5e-3, 5e-3, 7.2e-3, 8e-3, 9e-3,
       1e-2, 2e-2, 3e-2, 4e-2, 5e-2, 6e-2, 7e-2, 7.5e-2, 8e-2, 9e-2, 
       1e-1, 1.1e-1, 1.2e-1, 1.4e-1, 1.5e-1, 1.6e-1, 1.8e-1, 2e-1]
L = [10, 50, 100, 500, 1000, 5000, 10000]

delta = np.zeros((len(Amp), len(L)))
nonlin = np.zeros((len(Amp), len(L)))
NBRDT = np.zeros((len(Amp), len(L)))
ZF = np.zeros((len(Amp), len(L)))
for ia, amp in enumerate(Amp):
    for il, l in enumerate(L):
        tin = ((2*eds.a_today*l)/(3*(eds.t_today**(2/3))*R))**(3) 
        delt, delt1 = get_Amp1(tin, amp, l, R)
        delta[ia, il] = delt
        nonlin[ia, il] = abs(delt / delt1 - 1)
        aF = eds.a(tin)*1.68/delt
        tF = eds.t_func_a(aF)
        if tF > tin and eds.z(tF)>0:
            simtime = tF - tin
            courant_fac = 0.1
            N = 128
            dx = l / N
            dt = courant_fac*np.sqrt(3)*eds.a(tin)*dx
            NBRDT[ia, il] = simtime / dt
            ZF[ia, il] = eds.z(tF)
        else:
            NBRDT[ia, il] = -1
            ZF[ia, il] = -1
```

```python
def get_Amp(Amp):
    R = 10
    L = 1000
    N = 32
    tIN = ((2*eds.a_today*L)/(3*(eds.t_today**(2/3))*R))**(3) 
    params = {'Amp_pertx': Amp, 'Amp_perty': Amp, 'Amp_pertz': Amp, 
          'lambda_pertx':L, 'lambda_perty':L, 'lambda_pertz':L}
    output = A.f_Amp(tIN, [L]*3, N, params, inhom='sin', expansion='EdS')
    nonlin = abs(output['deltaOD']/output['delta1OD'] - 1)
    return nonlin
def get_delt(Amp):
    R = 10
    L = 1000
    N = 32
    tIN = ((2*eds.a_today*L)/(3*(eds.t_today**(2/3))*R))**(3) 
    params = {'Amp_pertx': Amp, 'Amp_perty': Amp, 'Amp_pertz': Amp, 
          'lambda_pertx':L, 'lambda_perty':L, 'lambda_pertz':L}
    output = A.f_Amp(tIN, [L]*3, N, params, inhom='sin', expansion='EdS')
    return output['deltaOD']
```

```python
def get_L(L):
    R = 10
    delt = 0.003008200954052631
    N = 128
    courant_fac = 0.1
    tIN = ((2*eds.a_today*L)/(3*(eds.t_today**(2/3))*R))**(3) 
    aF = eds.a(tIN)*1.68/delt
    tF = eds.t_func_a(aF)
    dx = L / N
    dt = courant_fac*np.sqrt(3)*eds.a(tIN)*dx
    return (tF - tin) / dt
def get_z(L):
    R = 10
    delt = 0.003008200954052631
    N = 128
    courant_fac = 0.1
    tIN = ((2*eds.a_today*L)/(3*(eds.t_today**(2/3))*R))**(3) 
    aF = eds.a(tIN)*1.68/delt
    tF = eds.t_func_a(aF)
    return eds.t_func_a(aF)

def get_zIN(L):
    R = 10
    delt = 0.003008200954052631
    N = 128
    courant_fac = 0.1
    tIN = ((2*eds.a_today*L)/(3*(eds.t_today**(2/3))*R))**(3)
    return tIN
```

```python
L=np.array([1e2, 2e2, 3e2, 4e2, 4.2e2, 4.4e2, 4.6e2, 4.8e2, 5e2, 6e2, 7e2, 8e2, 9e2, 
            1e3, 1.2e3, 1.4e3, 1.6e3, 1.8e3, 2e3, 3e3, 4e3, 5e3, 6e3, 7e3, 8e3, 9e3, 1e4])
plt.subplot(131)
plt.loglog(L, get_L(L))
plt.xlabel('L')
plt.ylabel(r'$N_{\Delta t}$')
plt.grid()
plt.subplot(132)
plt.loglog(L, get_z(L))
plt.loglog(L, get_zIN(L))
plt.xlabel('L')
plt.ylabel(r'$z$')
plt.grid()
plt.subplot(133)
plt.loglog(L, eds.z(get_z(L)))
plt.loglog(L, eds.z(get_zIN(L)))
plt.xlabel('L')
plt.ylabel(r'$z$')
plt.grid()
```

```python
# in my sim
TINnonl = 0.03366724282819571 #1
aINnonl = eds.a(TINnonl)
dINnonl = 0.003008200954052631 #0.03

# for eta = 0
TIN = 0.0001
aIN = eds.a(TIN)
dIN = (5 * dINnonl * aIN) / (3 * aINnonl)

# turn around
TTA = (3*np.pi/4)*TIN*(1+dIN)*(dIN**(-3/2))
aTA = eds.a(TTA)
etaTA = np.pi
d1TA = (3/5) * dIN * ((TTA/TIN)**(2/3))

# final
TF = 2*TTA
aF = eds.a(TF)
etaF = 2*np.pi
d1F = (3/5) * dIN * ((TF/TIN)**(2/3))

# rescale
def get_eta(eettaa):
    tt = TTA*(eettaa-np.sin(eettaa))/np.pi
    return tt
etaINnonl = num.dichotomy(TINnonl, get_eta, 0.1, 2.5*np.pi, 1e-10)
deltaINnonl = ((9*((etaINnonl - np.sin(etaINnonl))**2)) / (2*((1-np.cos(etaINnonl))**3))) - 1

# function of time
eta = np.arange(0.1, 2*np.pi, 0.01)
t = TTA*(eta-np.sin(eta))/np.pi
delta = (dINnonl/deltaINnonl) * (((9*((eta - np.sin(eta))**2)) / (2*((1-np.cos(eta))**3))) - 1)
delta1 = dINnonl * ((t/TINnonl)**(2/3))


plt.figure()
plt.semilogy(eta, delta)
plt.semilogy(eta, delta1)
plt.axvline(x = etaINnonl)
plt.axhline(y = dINnonl)
plt.axhline(y = d1TA, color='k')
plt.axvline(x = etaTA, color='k')
plt.axhline(y = d1F, color='g')
plt.axvline(x = etaF, color='g')
plt.ylim(0.001, 200)

plt.figure()
plt.semilogy(eta, eds.a(t)/aIN)
plt.semilogy(eta, eds.a(t)/aIN)
plt.axhline(y = eds.a(TTA)/aIN, color='k') # correct
plt.axvline(x = etaTA, color='k') # correct
plt.axhline(y = eds.a(TF)/aIN) # correct
plt.axvline(x = etaF) # correct
```
