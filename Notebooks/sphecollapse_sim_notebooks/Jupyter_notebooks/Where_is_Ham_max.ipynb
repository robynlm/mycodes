```python
def plot3d(data):
    cmap = matplotlib.cm.get_cmap('inferno')
    nbr_plots = 300
    isovals = np.linspace(np.min(data)*0.99, np.max(data)*0.99, nbr_plots)[::-1]
    iso = int(0.05*nbr_plots)
    m = matplotlib.cm.ScalarMappable(cmap=cmap)
    m.set_array(isovals)
    ticks = [0, (Lin.N-1)/4, (Lin.N-1)/2, 3*(Lin.N-1)/4, (Lin.N-1)]
    tickm = [-0.5, -0.25, 0.0, 0.25, 0.5]
    fig = plt.figure(figsize=(9, 6))
    gs = matplotlib.gridspec.GridSpec(1, 1) 
    ax = fig.add_subplot(gs[0], projection='3d')
    verts, faces, _, _ = measure.marching_cubes_lewiner(data, level=isovals[iso])
    ax.plot_trisurf(verts[:, 0], verts[:,1], faces, verts[:, 2], linewidth=0, alpha=1, edgecolor='none', antialiased=False, color=cmap(1-iso/nbr_plots))
    ax.set_xlim(0, Lin.N-1)
    ax.set_ylim(0, Lin.N-1)
    ax.set_zlim(0, Lin.N-1)
    ax.set_xticks(ticks)
    ax.set_xticklabels(tickm)
    ax.set_xlabel(r'$x/L$')
    ax.set_yticks(ticks)
    ax.set_yticklabels(tickm)
    ax.set_ylabel(r'$y/L$')
    ax.set_zticks(ticks)
    ax.set_zticklabels(tickm)
    ax.set_zlabel(r'$z/L$')
    ax.view_init(30,30)
```

```python
for it in np.arange(0,24300,100):
    it = int(it)
    f = h5py.File('{}_it_{:06d}.hdf5'.format(OGpath+'all_iterations/'+sim.sim_name, it), 'r')
    RicciS           = GetV.get_the_ricci(f, it)['RicciS']
    K_dict           = GetV.get_the_curv(f,it)
    rho              = GetV.get_the_rho(f, it)['rho']
    Ham              = RicciS + (2/3)*K_dict['K']**2 - 2*K_dict['A2'] - 2*Lin.kappa*rho - 2*Lin.Lambda
    Ham_energy_scale = np.sqrt(abs((RicciS)**2 + ((2/3)*K_dict['K']**2)**2 + (2*K_dict['A2'])**2 + (2*Lin.kappa*rho)**2 + (2*Lin.Lambda)**2))
    Hamn = Ham/Ham_energy_scale
    del RicciS, K_dict, rho, Ham, Ham_energy_scale
    print(it, Hamn[48,48,48], Hamn[np.where(abs(Hamn)==np.max(abs(Hamn)))], np.where(abs(Hamn)==np.max(abs(Hamn))))
    #plot3d(Hamn)
```
