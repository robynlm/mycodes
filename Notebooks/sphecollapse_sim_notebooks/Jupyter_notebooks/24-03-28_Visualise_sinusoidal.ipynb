```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import gridspec
```

```python
L = 1206
N = 128
dx = L/N
x1d = np.linspace(-L/2, L/2, N)
```

```python
x, y = np.meshgrid(x1d, x1d)
```

```python
f1 = np.sin(x*2*np.pi/L)
f2 = np.sin(y*2*np.pi/L)
```

```python
fs = 30
def plot(f):
    fig = plt.figure(figsize=(10,10))
    gs = gridspec.GridSpec(1, 1) 
    ax = fig.add_subplot(gs[0])
    plt.imshow(f, cmap=matplotlib.cm.get_cmap('inferno', 9))
    ticks = [0, N/4, N/2, 3*N/4, N-1]
    tickm = [-0.5, -0.25, 0.0, 0.25, 0.5]
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.set_xticklabels(tickm, fontsize=fs*0.8)
    ax.set_yticklabels(tickm, fontsize=fs*0.8)
    ax.set_xlabel(r'$x \;\; [\lambda_{pert}]$', fontsize=fs)
    ax.set_ylabel(r'$y \;\; [\lambda_{pert}]$', fontsize=fs)
plot(f1)
plot(f2)
plot(f1+f2)
```
