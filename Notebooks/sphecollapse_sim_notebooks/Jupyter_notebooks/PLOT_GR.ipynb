```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ygrav(x, y):
    Egrav_impact = 360/np.sqrt((35-x)**2+(275-y)**2)**0.5
    Mgrav_impact = 170/np.sqrt((75-x)**2+(400-y)**2)**0.5
    xscale = 0
    yscale = 1
    plt.plot(x-xscale*(Egrav_impact+Mgrav_impact), y-yscale*(Egrav_impact+Mgrav_impact), color='xkcd:bluish')
    
def ygravN(x, y):
    plt.plot(x, y, color='xkcd:bluish')
```

```python
x = np.arange(0, 100, 0.25)
yaigu = x*np.sin((60)*np.pi/180)
yobtu = -x*np.sin((60)*np.pi/180)
nbr_lines = 100
dy = 10
plt.figure(figsize=(10,10))
#nbr_stars = 30
#plt.scatter(100*np.random.random(nbr_stars), 425*np.random.random(nbr_stars)+75, s=75*np.random.random(nbr_stars), marker='*', color='w')
for i in range(nbr_lines):
    ygrav(x, yaigu+i*dy)
    ygrav(x, yobtu+i*dy)
plt.gca().set_facecolor('xkcd:deep blue')
plt.xlim(0, 100)
plt.ylim(-10, 350)
#plt.ylim(75, 75+175*2)
#plt.gca().set_yticks([])
plt.gca().set_xticks([])
plt.savefig('/home/robynm/Documents/poster/P-Newt')
```

```python
from mpl_toolkits.mplot3d import Axes3D

dx = 20
x = np.arange(-200, 800, dx)
x, y =np.meshgrid(x,x)
Egrav_impact = 360/np.sqrt((175-x)**2+(175-y)**2)**0.5
Mgrav_impact = 170/np.sqrt((350-x)**2+(350-y)**2)**0.5
z = np.zeros(np.shape(x))-(Egrav_impact+Mgrav_impact)

fig = plt.figure(figsize=(40,40))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x,y,z)
ax.view_init(50,-70)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_facecolor('xkcd:deep blue')
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.set_axis_off()
plt.savefig('/home/robynm/Documents/poster/P-GR')
```
