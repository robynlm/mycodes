```python
allIT = np.arange(0,48500+400,400)
```

```python
def abreal(a):
    return abs(np.real(a))
def abim(a):
    return abs(np.imag(a))
```

```python
def Petrov_determination(it, invar_dict, cutoff):
    H = Lin.temp_from_temp('H', 'it', it)
    #HK = -K/3
    E2 = np.array(invar_dict['E2'])
    B2 = np.array(invar_dict['B2'])
    dE = abs(E2)**(1/2)
    dB = abs(B2)**(1/2)
    dEB = abs(E2 - B2)**(1/2)
    
    I = np.array(invar_dict['I'])
    J = np.array(invar_dict['J'])
    dI = I**(1/2)
    dJ = J**(1/3)
    
    Scrit = I*I*I-27*J*J
    dScrit = Scrit**(1/6)
    
    dLS = np.array(invar_dict['LS'])**(1/2)
    dK = np.array(invar_dict['K'])**(1/3)
    dN = np.array(invar_dict['N'])**(1/4)
    
    Petrov_class = np.copy(E2)
    cadjust = 0.5
    for i in range(Lin.N):
        for j in range(Lin.N):
            #H = HK[i,j]
            cutoff = 0.1
            Scriteq0 = abreal(dScrit[i,j])<cutoff*H*H and abim(dScrit[i,j])<cutoff*H*H
            if Scriteq0:
                cutoff *= cadjust
                Ieq0 = abreal(dI[i,j])<cutoff*H*H and abim(dI[i,j])<cutoff*H*H
                Jeq0 = abreal(dJ[i,j])<cutoff*H*H and abim(dJ[i,j])<cutoff*H*H
                if Ieq0 and Jeq0: # and dEB[i,j]<cutoff*H*H:
                    cutoff *= cadjust
                    Keq0 = abreal(dK[i,j])<cutoff*H*H and abim(dK[i,j])<cutoff*H*H
                    LSeq0 = abreal(dLS[i,j])<cutoff*H*H and abim(dLS[i,j])<cutoff*H*H
                    if Keq0 and LSeq0:
                        cutoff *= cadjust
                        if dE[i,j]<cutoff*H*H and dB[i,j]<cutoff*H*H:
                            Petrov_class[i,j]=6 #type O
                        else:
                            Petrov_class[i,j]=5 #type N
                    else:
                        Petrov_class[i,j]=4 #type III
                else:
                    Keq0 = abreal(dK[i,j])<cutoff*H*H and abim(dK[i,j])<cutoff*H*H
                    Neq0 = abreal(dN[i,j])<cutoff*H*H and abim(dN[i,j])<cutoff*H*H
                    if Keq0 and Neq0:
                        Petrov_class[i,j]=3 #type D
                    else:
                        Petrov_class[i,j]=2 #type II
            else:
                Petrov_class[i,j]=1 #type I
    return Petrov_class
```

```python
for it in allIT:
    cutoff = 1e-1
    invar_dict = h5py.File(invar_path+'invar_diag_{:06d}.hdf5'.format(it), 'r')
    K = np.array(h5py.File(invar_path+'K_diag_{:06d}.hdf5'.format(it), 'r')['K'])
    Petrov_class = Petrov_determination(it, invar_dict, cutoff)
                
    fig, ax = plt.subplots(figsize=(6,6))
    cmap = plt.cm.get_cmap('BuPu', 6)
    plt.pcolormesh(Lin.d3xyz/sim.L, Lin.d3xyz*np.sqrt(2)/sim.L, Petrov_class, cmap=cmap, vmin=0.5, vmax=6.5, shading='nearest')
    plt.gca().set_aspect("equal")
    cbar = plt.colorbar(extend='both', ticks=[1, 2, 3, 4, 5, 6])
    cbar.ax.set_yticklabels(['I', 'II', 'D', 'III', 'N', 'O'])
    plt.title(r'$a/a_{IN}=$'+'{:.1f}, '.format(Lin.temp_from_temp('an', 'it', it))+r'$c=1e-$'+str(int(np.round(abs(np.log(cutoff)/np.log(10)), 1))))
    plt.ylabel('d'+r'$\;\;[\lambda_{pert}]$')
    plt.xlabel('x'+r'$\;\;[\lambda_{pert}]$')
    #plt.xlim(-0.5, 0)
    #plt.ylim(-0.7, 0)
    plt.savefig('/home/robynm/simulations/'+sim.sim_name+'/plots/Petrov/Petrov_diag_{:.0e}_it={:06d}.jpg'.format(cutoff, it))
```
