```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.6737
c = 1
H0 = h*c/2997.9
a0 = 1
t0EdS = 2/(3*H0)
Omega_m0 = 0.3147
Omega_l0 = 1 - Omega_m0
```

```python
# EdS
def a(t):
    return a0*(t/t0EdS)**(2/3)
```

```python
# LCDM
def a(t):
    return a0*(Omega_m0/Omega_l0)**(1/3)*np.sinh(np.sqrt(Omega_l0)*t/t0EdS)**(2/3)
```

```python
tIN = 1
tF = 1000
tC = 500
```

```python
L = 1821
N = 128
dx = L/N
courant_fac = 1/10
dtfac = a(tC)
```

```python
t = tIN
nbr_steps = 0
while t<tF:
    courant_minv = 3*a(t)*dx
    dtv = courant_fac*courant_minv/np.sqrt(3)
    
    courant_minc = 3*dtfac*dx
    dtc = courant_fac*courant_minc/np.sqrt(3)
    
    if dtv<dtc:
        dt = dtv
        print('{:.3f}, {:.3f}, {:.3f}'.format(t, dtv, dtc), ' varying')
    else:
        dt = dtc
        print('{:.3f}, {:.3f}, {:.3f}'.format(t, dtv, dtc), ' constant')
        
    t += dt
    nbr_steps += 1
```
