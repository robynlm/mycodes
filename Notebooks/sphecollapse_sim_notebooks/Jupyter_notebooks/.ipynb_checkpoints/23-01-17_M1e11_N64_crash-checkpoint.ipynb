```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from data_analysis_codes import param
from data_analysis_codes.tools import Cstyle
from data_analysis_codes.tools import LinData
plt.style.use(Cstyle.style1)
from data_analysis_codes.tools import ReadingTools as RRead
import h5py
```

```python
param = RRead.read_parameters('pflrw_M1e11_L3_zIN300_zC20_N64_LCDM_TV')
Lin = LinData.LinData_Class(param)
c = pd.read_table(param['datapath']+'constraints.csv', delimiter=',')
f = pd.read_table(param['datapath']+'h5_data.csv', delimiter=',')
```

```python
locs = ['maxabs', 'highQabs', 'medianabs', 'lowQabs', 'minabs']
loclabs = ['max', 'first quartile', 'median', 'third quartile', 'min']
cols = plt.cm.viridis(np.linspace(0,1,len(locs)))

plt.figure(figsize=(12, 5))
plt.subplot(121)
for i in range(5):
    plt.semilogy(Lin.an_initial(c['t']), c['Ham/HamEScale_'+locs[i]], color=cols[i], label=loclabs[i])
plt.grid()
plt.ylim(2e-1, 2e0)
plt.xlabel(r'$a/a_{IN}$')
plt.title(r'$|\mathcal{H}/[\mathcal{H}]|$')

plt.subplot(122)
for i in range(5):
    plt.semilogy(Lin.an_initial(c['t']), c['Mom1/MomEScale_'+locs[i]], color=cols[i], label=loclabs[i])
plt.grid()
plt.legend(bbox_to_anchor=(1,1))
plt.ylim(2e-8, 1e-1)
plt.xlabel(r'$a/a_{IN}$')
plt.title(r'$|\mathcal{M}_{1}/[\mathcal{M}]|$')
```

```python
def delta(t, loc):
    return f['drho_'+loc][0] * Lin.an_initial(t)

def drho(t, loc):
    return delta(t, loc)

def dK(t, loc):
    return - Lin.evo.fL(t) * delta(t, loc) / 3
    
def dgdet(t, loc):
    return - 6 * ( (delta(t, loc) / 3) + Lin.Rc(loc=loc))
```

```python
an = Lin.an_initial(f['t'])

plt.figure(figsize=(12, 5))
plt.subplot(131)
plt.plot(an, f['drho_OD'], color='C1', label='OD')
plt.plot(an, f['drho_cent'], color='C2', label='cent')
plt.plot(an, f['drho_UD'], color='C0', label='UD')
plt.plot(an, drho(f['t'], 'OD'), linestyle=':', color='firebrick', label='OD'+r'${}^{(1)}$')
plt.plot(an, drho(f['t'], 'cent'), linestyle=':', color='gold', label='cent'+r'${}^{(1)}$')
plt.plot(an, drho(f['t'], 'UD'), linestyle=':', color='skyblue', label='UD'+r'${}^{(1)}$')
xmin, xmax = plt.gca().axes.get_xlim()
x = np.arange(xmin-10, xmax+10, 10)
plt.fill_between(x, [-10]*len(x), [-1]*len(x), facecolor='brown', edgecolor=None, alpha=0.2, label='Unphysical')
plt.xlim(xmin, xmax)
plt.text(1, -1.5, 'Unphysical \n region', c='brown')
plt.grid()
plt.ylim(-1.75, 2)
plt.xlabel(r'$a/a_{IN}$')
plt.title(r'$\delta$')

plt.subplot(132)
plt.plot(an, f['dgdet_OD'], color='C1', label='OD')
plt.plot(an, f['dgdet_cent'], color='C2', label='cent')
plt.plot(an, f['dgdet_UD'], color='C0', label='UD')
plt.plot(an, dgdet(f['t'], 'OD'), linestyle=':', color='firebrick', label='OD'+r'${}^{(1)}$')
plt.plot(an, dgdet(f['t'], 'cent'), linestyle=':', color='gold', label='cent'+r'${}^{(1)}$')
plt.plot(an, dgdet(f['t'], 'UD'), linestyle=':', color='skyblue', label='UD'+r'${}^{(1)}$')
xmin, xmax = plt.gca().axes.get_xlim()
x = np.arange(xmin-10, xmax+10, 10)
plt.fill_between(x, [-10]*len(x), [-1]*len(x), facecolor='brown', edgecolor=None, alpha=0.2, label='Unphysical')
plt.xlim(xmin, xmax)
plt.text(1, -1.5, 'Unphysical \n region', c='brown')
plt.grid()
plt.ylim(-1.75, 2)
plt.xlabel(r'$a/a_{IN}$')
plt.title(r'$\delta\gamma$')

plt.subplot(133)
plt.plot(an, f['dK_OD'], color='C1', label='OD')
plt.plot(an, f['dK_cent'], color='C2', label='cent')
plt.plot(an, f['dK_UD'], color='C0', label='UD')
plt.plot(an, dK(f['t'], 'OD'), linestyle=':', color='firebrick', label='OD'+r'${}^{(1)}$')
plt.plot(an, dK(f['t'], 'cent'), linestyle=':', color='gold', label='cent'+r'${}^{(1)}$')
plt.plot(an, dK(f['t'], 'UD'), linestyle=':', color='skyblue', label='UD'+r'${}^{(1)}$')
#plt.axhline(y=-1, c='brown')
#plt.text(1, -0.98, 'Turn around', c='brown')
#plt.axhline(y=0.5, c='brown')
#plt.text(1, 0.52, 'Milne model', c='brown')
plt.grid()
plt.ylim(-0.4, 0.2)
plt.legend(bbox_to_anchor=(1,1))
plt.xlabel(r'$a/a_{IN}$')
plt.title(r'$\delta K$')
```

```python
k = 2*np.pi/param['ICPertFLRW_lambda_x[0]']
Rc = param['ICPertFLRW_Amp_x[0]']*(np.sin(k*Lin.d3x)+np.sin(k*Lin.d3y)+np.sin(k*Lin.d3z))
dxdxRc = -param['ICPertFLRW_Amp_x[0]']*k*k*np.sin(k*Lin.d3x)
```

```python
t = param['cctk_initial_time']
F = Lin.evo.fL(t)+3*Lin.evo.Omega_m(t)/2
gxx = Lin.evo.a(t)*Lin.evo.a(t)*(1-2*Rc) - (2*dxdxRc/(F*Lin.evo.Hprop(t)*Lin.evo.Hprop(t)))
```

```python
diff = d['ADMBASE::gxx it=0 tl=0 rl=0']-gxx
plt.imshow(diff[:,:,16])
plt.colorbar()
```

```python
data_delta = d['CT_DUST::rho it=0 tl=0 rl=0']/Lin.evo.rho(t)-1
delta1 = -(k**2)*Rc/ (F*((Lin.evo.a(t)*Lin.evo.Hprop(t))**2))
```

```python
diff = data_delta-delta1
plt.imshow(diff[:,:,16])
plt.colorbar()
```
