```python
def interpolate(f):
    f_save = [f[0]]
    for i in range(len(f)-1):
        f_save += [np.average([f[i], f[i+1]]), f[i+1]]
    return np.array(f_save)

def extrapolate(f):
    return np.append(f, f[-1]+f[-1]-f[-2])

def get_error(f32, f64, f128):
    f32 = interpolate(interpolate(f32))
    f64 = interpolate(f64)
    while len(f32)<len(f128):
        f32 = extrapolate(f32)
    while len(f64)<len(f128):
        f64 = extrapolate(f64)
    c = abs(f32-f64)/abs(f64-f128)
    return abs(f64-f128)/abs(c-1)
```

```python
loc = 'av'
an = Lin128.an_initial(np.array(f128['t']))
H = Lin128.Hprop(np.array(f128['t']))
H32 = Lin128.Hprop(np.array(f32['t']))
H64 = Lin128.Hprop(np.array(f64['t']))

plt.subplot(211)
plt.semilogy(an, np.sqrt(f128['B2_'+loc])/H**2, 'C0-', label=r'$\langle|B|\rangle_{\mathcal{D}}/H^2$')
err = get_error(np.sqrt(f32['B2_'+loc])/H32**2, np.sqrt(f64['B2_'+loc])/H64**2, np.sqrt(f128['B2_'+loc])/H**2)
plt.fill_between(an, np.sqrt(f128['B2_'+loc])/H**2-err, np.sqrt(f128['B2_'+loc])/H**2+err, color='C0', alpha=0.15)

plt.semilogy(an, f128['divB_norm_'+loc]/H**3, 'C1-', label=r'$\langle|\nabla \cdot B|\rangle_{\mathcal{D}}/H^3$')
err = get_error(f32['divB_norm_'+loc]/H32**3, f64['divB_norm_'+loc]/H64**3, f128['divB_norm_'+loc]/H**3)
plt.fill_between(an, f128['divB_norm_'+loc]/H**3-err, f128['divB_norm_'+loc]/H**3+err, color='C1', alpha=0.15)

plt.semilogy(an, f128['curlB_norm_'+loc]/H**3, 'C2-', label=r'$\langle|\nabla \times B|\rangle_{\mathcal{D}}/H^3$')
err = get_error(f32['curlB_norm_'+loc]/H32**3, f64['curlB_norm_'+loc]/H64**3, f128['curlB_norm_'+loc]/H**3)
plt.fill_between(an, f128['curlB_norm_'+loc]/H**3-err, f128['curlB_norm_'+loc]/H**3+err, color='C2', alpha=0.15)
plt.xticks([10, 20, 30, 40, 50], [])
plt.grid()
plt.legend(bbox_to_anchor=(-0.15,1.1))

plt.subplot(212)
plt.semilogy(an, np.sqrt(f128['E2_'+loc])/H**2, 'C0-', label=r'$\langle|E|\rangle_{\mathcal{D}}/H^2$')
err = get_error(np.sqrt(f32['E2_'+loc])/H32**2, np.sqrt(f64['E2_'+loc])/H64**2, np.sqrt(f128['E2_'+loc])/H**2)
plt.fill_between(an, np.sqrt(f128['E2_'+loc])/H**2-err, np.sqrt(f128['E2_'+loc])/H**2+err, color='C0', alpha=0.15)

plt.semilogy(an, f128['divE_norm_'+loc]/H**3, 'C1-', label=r'$\langle|\nabla \cdot E|\rangle_{\mathcal{D}}/H^3$')
err = get_error(f32['divE_norm_'+loc]/H32**3, f64['divE_norm_'+loc]/H64**3, f128['divE_norm_'+loc]/H**3)
plt.fill_between(an, f128['divE_norm_'+loc]/H**3-err, f128['divE_norm_'+loc]/H**3+err, color='C1', alpha=0.15)

plt.semilogy(an, f128['curlE_norm_'+loc]/H**3, 'C2-', label=r'$\langle|\nabla \times E|\rangle_{\mathcal{D}}/H^3$')
err = get_error(f32['curlE_norm_'+loc]/H32**3, f64['curlE_norm_'+loc]/H64**3, f128['curlE_norm_'+loc]/H**3)
plt.fill_between(an, f128['curlE_norm_'+loc]/H**3-err, f128['curlE_norm_'+loc]/H**3+err, color='C2', alpha=0.15)
plt.xticks([10, 20, 30, 40, 50], ['10', '20', '30', '40', '50'])
plt.grid()
plt.legend(bbox_to_anchor=(-0.15,1.1))
plt.xlabel(r'$a/a_{IN}$')
```
