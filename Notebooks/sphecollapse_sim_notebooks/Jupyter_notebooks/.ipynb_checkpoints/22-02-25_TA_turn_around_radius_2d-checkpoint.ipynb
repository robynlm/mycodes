```python
import numpy as np
import matplotlib.pyplot as plt
from data_analysis_codes.tools import PlottingTools as RPlot
from data_analysis_codes.tools import ReadingTools as RRead
from data_analysis_codes import param
from data_analysis_codes.tools import Cstyle
from data_analysis_codes.tools import LinData
plt.style.use(Cstyle.style1)
import time
import h5py
from matplotlib import gridspec
from data_analysis_codes.tools import ODUDLoc
import pandas as pd
from data_analysis_codes.tools import TAradius
from data_analysis_codes.tools import GetVars_Plot2d as GVar
from data_analysis_codes.tools import Plot2d
```

```python
sim = param.pflrw_d3e2_L1821_t1_N128_LCDM()
path = '/home/robynm/simulations/'+sim.sim_name+'/'
sim128 = RRead.read_parameters('pflrw_d3e2_L1821_t1_N128_LCDM')
Lin = LinData.LinData_Class(sim128)
OGpath = '/home/robynm/simulations/'+sim.sim_name+'/output-0000/pflrw_L1821/'
get_var = GVar.Get_var(sim128, Lin)
makeplot = Plot2d.MyPlot2dclass(sim128, Lin, logcolor=True)
itoptions = [0, 5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000]
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
it = 30000
f = h5py.File('{}_it_{:06d}.hdf5'.format(sim128['h5datapath']+sim.sim_name, it), 'r')
K = get_var.get_the_curv(f, it)['K']
makeplot.f_xy(K*sim.L, it)
plt.plot(Lin.d3xyz/sim.L, [-1/4]*len(Lin.d3xyz), 'w-')
plt.plot([-1/4]*len(Lin.d3xyz), Lin.d3xyz/sim.L, 'w-')
plt.plot(Lin.d3xyz/sim.L, Lin.d3xyz/sim.L, 'w-.')
plt.plot(Lin.d3xyz/sim.L, -0.5-Lin.d3xyz/sim.L, 'w-.')
plt.xlim(-0.5, (sim.L/2-sim.dx)/sim.L)
plt.ylim(-0.5, (sim.L/2-sim.dx)/sim.L)
an = Lin.temp_from_temp('an', 'it', it)
plt.title(r'$|\Theta| \;\; [\lambda_{pert}^{-1}] \;\; $'+'\n at'
          +'$ \;\; a/a_{IN}=$'+'{:.2f}'.format(an), fontsize = 20)

plt.subplot(122)
it = 45000
f = h5py.File('{}_it_{:06d}.hdf5'.format(sim128['h5datapath']+sim.sim_name, it), 'r')
K = get_var.get_the_curv(f, it)['K']
makeplot.f_xy(K*sim.L, it)
plt.plot(Lin.d3xyz/sim.L, [-1/4]*len(Lin.d3xyz), 'w-')
plt.plot([-1/4]*len(Lin.d3xyz), Lin.d3xyz/sim.L, 'w-')
plt.plot(Lin.d3xyz/sim.L, Lin.d3xyz/sim.L, 'w-.')
plt.plot(Lin.d3xyz/sim.L, -0.5-Lin.d3xyz/sim.L, 'w-.')
plt.xlim(-0.5, (sim.L/2-sim.dx)/sim.L)
plt.ylim(-0.5, (sim.L/2-sim.dx)/sim.L)
an = Lin.temp_from_temp('an', 'it', it)
plt.title(r'$|\Theta| \;\; [\lambda_{pert}^{-1}] \;\; $'+'\n at'
          +'$ \;\; a/a_{IN}=$'+'{:.2f}'.format(an), fontsize = 20)

plt.tight_layout()
plt.savefig('/home/robynm/simulations/'+sim.sim_name+'/plots/TA_shape.pdf')
```
