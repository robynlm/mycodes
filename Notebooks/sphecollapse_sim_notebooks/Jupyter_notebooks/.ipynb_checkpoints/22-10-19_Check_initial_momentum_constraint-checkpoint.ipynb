```python
import numpy as np
import data_analysis_codes.initialconditions as ic
from data_analysis_codes.ebweyl import ebweyl as ebw
import matplotlib.pyplot as plt
```

```python
def av(phi, sgdet, dx):
    V = np.sum(sgdet)*(dx**3)
    return np.sum(phi*sgdet)*(dx**3)/V
```

```python
errM1_max, errM2_max, errM3_max = [], [], []
errM1_av1, errM2_av1, errM3_av1 = [], [], []
errM1_av2, errM2_av2, errM3_av2 = [], [], []
Ns = [16, 32, 64]
for N in Ns:
    L = 1821
    dx = L/N
    IC = ic.ICPertFLRW(1e-3, 1, L, N, 'LCDM')
    FD = ebw.FiniteDifference(dx, N)
    EBW = ebw.Weyl(FD, IC.gdown4, IC.Kdown4)
    Gudd3 = EBW.christoffel_symbol_udd3()
    
    sgdet = np.sqrt(ebw.determinant3(EBW.gammadown3))
    
    DdKdd = EBW.covariant_derivatice_3_tensor2down3(Gudd3, EBW.Kdown3)
    DdKud = np.einsum('bc..., bca... -> a...', EBW.gammaup3, DdKdd)
    DdK = np.einsum('bc..., acb... -> a...', EBW.gammaup3, DdKdd)
    Mom = DdKud - DdK
    
    DdKud_DdKuu = np.einsum('a..., ad..., d... -> ...', DdKud, EBW.gammaup3, DdKud)
    DdK_DuK = np.einsum('a..., ad..., d... -> ...', DdK, EBW.gammaup3, DdK)
    Mom_rel = DdKud_DdKuu - DdK_DuK
    
    errM1_max += [np.max(Mom[0]/Mom_rel)]
    errM2_max += [np.max(Mom[1]/Mom_rel)]
    errM3_max += [np.max(Mom[2]/Mom_rel)]
    
    errM1_av1 += [av(Mom[0]/Mom_rel, sgdet, dx)]
    errM2_av1 += [av(Mom[1]/Mom_rel, sgdet, dx)]
    errM3_av1 += [av(Mom[2]/Mom_rel, sgdet, dx)]
    
    av_Mom_rel = av(Mom_rel, sgdet, dx)
    errM1_av2 += [av(Mom[0], sgdet, dx)/av_Mom_rel]
    errM2_av2 += [av(Mom[1], sgdet, dx)/av_Mom_rel]
    errM3_av2 += [av(Mom[2], sgdet, dx)/av_Mom_rel]
```

```python
plt.plot(Ns, errM1_av2, marker='o')
plt.plot(Ns, errM2_av2, marker='o')
plt.plot(Ns, errM3_av2, marker='o')
plt.grid()
```
