```python
import numpy as np
import matplotlib.pyplot as plt
from data_analysis_codes import param
from data_analysis_codes.tools import LinData
from data_analysis_codes.tools import ODUDLoc
from data_analysis_codes.tools import MassCalc
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
p = param.pflrw_d3e2_L1821_t1_N256_LCDM()
Lin = LinData.LinData_Class(p, '/home/robynm/simulations/pflrw_d3e2_L1821_t1_N256_LCDM/')
locfinder = ODUDLoc.ODUDLocClass(p)
MC = MassCalc.MassCalcClass(p, Lin, locfinder.findlocations()[0])
```

```python
xyz0 = Lin.d3xyz[locfinder.findlocations()[0][0]][0]
R = np.sqrt((Lin.d3x-xyz0)**2+(Lin.d3y-xyz0)**2+(Lin.d3z-xyz0)**2)
phi = R
plt.figure()
plt.imshow(phi[:,:,int(Lin.N/4)])

Box1 = np.ones(np.shape(phi))
Box0 = np.zeros(np.shape(phi))
gdet = (Lin.a_initial**6)*Box1
```
