```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import h5py
from data_analysis_codes.tools import LinData
from data_analysis_codes.tools import ReadingTools as RRead
from data_analysis_codes import param as paramf
from data_analysis_codes.tools import Plot2d
from data_analysis_codes.tools import TAradius
from data_analysis_codes.tools import ODUDLoc
from data_analysis_codes.tools import Cstyle
plt.style.use(Cstyle.style1)
from data_analysis_codes.tools import GetVars_Plot2d as GVar
```

```python
sim = paramf.pflrw_d3e2_L1821_t1_N32_LCDM()
param = RRead.read_parameters('pflrw_d3e2_L1821_t1_N32_LCDM')
Lin = LinData.LinData_Class(param)
PLT = Plot2d.MyPlot2dclass(param, Lin)
locfinder = ODUDLoc.ODUDLocClass(param)
defined_locs = locfinder.findlocations()
TAr = TAradius.TA_Class(param, Lin, defined_locs[0])
get_var = GVar.Get_var(param, Lin)
```

```python
it = 0
it_file_name = '/home/robynm/simulations/'+sim.sim_name+'/output-0000/pflrw_L1821/all_iterations/'+sim.sim_name
f = h5py.File('{}_it_{:06d}.hdf5'.format(it_file_name, it), 'r')
    
Kdict = get_var.get_the_curv(f, it)
gdet = Kdict['metric_dic']['gdet']
delta = get_var.get_the_rho(f, it)['drho']

data = delta
Data = []
for ix in range(param['Nx']):
    for iy in range(param['Nx']):
        #r, theta, phi = TAr.get_proper_Radius_of_grid_point(gdet, (ix, iy, iy))
        r, theta, phi = TAr.get_Radius_of_grid_point(gdet, (ix, iy, iy))
        r *=  Lin.temp_from_temp('a', 'it', it)
        x = r*np.cos(phi)*np.sin(theta)
        y = r*np.sin(phi)*np.sin(theta)
        z = r*np.cos(theta)
        d = y*np.sqrt(2)
        Data += [[data[(ix, iy, iy)], x, d]]
Data = np.array(Data)
Data = Data[Data[:, 0].argsort()]
    
plt.figure(figsize=(7,7), facecolor='white')
plt.scatter(Data[:,1], Data[:,2], c=Data[:,0])
plt.xlim(-3.1, 3.3)
plt.ylim(-4.3, 4.5)
plt.gca().set_aspect("equal")
plt.xlabel('x [Mpc]')
plt.ylabel('y = z [Mpc]')
plt.title(r'$\delta \;\; a/a_{IN}=$'+'{:.2f}'.format(Lin.temp_from_temp('an', 'it', it)))
plt.colorbar(format='%.1e', extend='both')
plt.savefig("/home/robynm/simulations/"+sim.sim_name+"/delta_scatter_1.png")
```

```python
for it in np.arange(0, 12200, 100):
    it_file_name = '/home/robynm/simulations/'+sim.sim_name+'/output-0000/pflrw_L1821/all_iterations/'+sim.sim_name
    f = h5py.File('{}_it_{:06d}.hdf5'.format(it_file_name, it), 'r')
    
    Kdict = get_var.get_the_curv(f, it)
    gdet = Kdict['metric_dic']['gdet']
    K = Kdict['K']
    #delta = get_var.get_the_curv(f, it)['drho']

    data = abs(K)*sim.L
    Data = []
    for ix in range(param['Nx']):
        for iy in range(param['Nx']):
            r, theta, phi = TAr.get_proper_Radius_of_grid_point(gdet, (ix, iy, iy))
            x = r*np.cos(phi)*np.sin(theta)
            y = r*np.sin(phi)*np.sin(theta)
            z = r*np.cos(theta)
            d = y*np.sqrt(2)
            Data += [[data[(ix, iy, iy)], x, d]]
    Data = np.array(Data)
    Data = Data[Data[:, 0].argsort()]
    
    plt.figure(figsize=(7,7), facecolor='white')
    plt.scatter(Data[:,1], Data[:,2], c=Data[:,0])
    plt.xlim(-150, 150)
    plt.ylim(-250, 250)
    plt.gca().set_aspect("equal")
    plt.xlabel('x [Mpc]')
    plt.ylabel('y = z [Mpc]')
    plt.title(r'$|\Theta| \;\; [\lambda_{pert}^{-1}] \;\; a/a_{IN}=$'+'{:.2f}'.format(Lin.temp_from_temp('an', 'it', it)))
    plt.colorbar(format='%.1e', extend='both')
    #plt.tight_layout(pad=2)
    plt.savefig("/home/robynm/simulations/"+sim.sim_name+"/Collapse_video4/it=%06d.png" % it)
```
