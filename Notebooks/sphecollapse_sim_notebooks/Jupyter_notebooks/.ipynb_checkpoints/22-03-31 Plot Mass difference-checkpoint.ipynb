```python
sim16 = param.pflrw_d3e2_L1821_t1_N16_LCDM()
Lin16 = LinData.LinData_Class(sim16, '/home/robynm/simulations/'+sim16.sim_name+'/')
f16 = pd.read_table('/home/robynm/simulations/'+sim16.sim_name+'/mass_radius_evo.csv', delimiter=',')

sim32 = param.pflrw_d3e2_L1821_t1_N32_LCDM()
Lin32 = LinData.LinData_Class(sim32, '/home/robynm/simulations/'+sim32.sim_name+'/')
f32 = pd.read_table('/home/robynm/simulations/'+sim32.sim_name+'/mass_radius_evo.csv', delimiter=',')

sim64 = param.pflrw_d3e2_L1821_t1_N64_LCDM()
Lin64 = LinData.LinData_Class(sim64, '/home/robynm/simulations/'+sim64.sim_name+'/')
f64 = pd.read_table('/home/robynm/simulations/'+sim64.sim_name+'/mass_radius_evo.csv', delimiter=',')

sim128 = param.pflrw_d3e2_L1821_t1_N128_LCDM()
Lin128 = LinData.LinData_Class(sim128, '/home/robynm/simulations/'+sim128.sim_name+'/')
f128 = pd.read_table('/home/robynm/simulations/'+sim128.sim_name+'/mass_radius_evo.csv', delimiter=',')
```

```python
radius_denominator = np.append(np.arange(2, 11, 1), np.arange(20, 128, 10))

cmap = plt.get_cmap('copper')
for enu, i in enumerate(radius_denominator):
    plt.semilogy(Lin128.an_initial(np.array(f128['t'])), abs((f128['Mass_'+str(i)]*Lin128.a(np.array(f128['t']))**3)/f128['Mass_prop_'+str(i)]), c=cmap(i/len(radius_denominator)))#,label=r'$R_{com}=\lambda_{pert}/$'+str(i))

sm = plt.cm.ScalarMappable(cmap=cmap, norm=mpl.colors.LogNorm(vmin=1821/120, vmax=1821/2))
cbar = plt.colorbar(sm, label=r'$R_{com}$')
cbar.set_ticks([1821/120, 1821/20, 1821/2])
cbar.set_ticklabels([r'$\lambda_{pert}/2$', r'$\lambda_{pert}/20$', r'$\lambda_{pert}/120$'])

plt.xlabel(r'$a/a_{IN}$')
plt.ylabel(r'$|M_{grav}/M_{prop}|$')
plt.grid()
```
