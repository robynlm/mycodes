```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 3e8
h = 0.6737
H0 = c*h / 2997.9
```

```python
t_rad = np.geomspace(1e-10, 47000, 100)
t_mat = np.geomspace(47000, 9.8e9, 100)
t_dar = np.geomspace(9.8e9, 13.8e9, 100)

a_rad = t_rad**(1/2) 
a_mat = t_mat**(2/3)
a_dar = np.exp(H0 * t_dar)

def d_dec(a):
    return a**(-3/2)
def d_gro(a):
    return a
```

```python
plt.figure()
plt.loglog(t_rad, d_dec(a_rad), label='decay')
plt.loglog(t_rad, d_gro(a_rad), label='grow')
plt.loglog(t_mat, d_dec(a_mat), 'C0')
plt.loglog(t_mat, d_gro(a_mat), 'C1')
plt.loglog(t_dar, d_dec(a_dar), 'C0')
plt.loglog(t_dar, d_gro(a_dar), 'C1')
plt.grid()
plt.legend()
plt.axvline(x = 47000)
plt.axvline(x = 9.8e9)
```
