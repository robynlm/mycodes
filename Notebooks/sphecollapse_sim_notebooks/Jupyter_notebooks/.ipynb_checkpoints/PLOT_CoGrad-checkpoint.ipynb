```python
from data_analysis_codes import param
from data_analysis_codes.tools import PlottingTools as RPlot

p  = param.pflrw_d3e2_L6_t1_N16_zc200()
pd = RPlot.getdata(p)
p0  = param.pflrw_d3e2_L1206_t1_N16_zc0()
pd0 = RPlot.getdata(p0)
```

```python
import matplotlib.pyplot as plt
from data_analysis_codes.tools import Cstyle
plt.style.use(Cstyle.style1)
from matplotlib import gridspec
from data_analysis_codes.tools import LinData
```

```python
def pformat(pnbr):
    ytfs=20
    xlab = r'$a/a_{IN}$'
    titles = {0:r'$|\mathcal{D}_i|$', 1:r'$|\mathcal{Z}_i|$', 2:r'$|\mathcal{C}_i|$', 3:'', 4:'', 5:''}
    plt.title(titles[pnbr])
    plt.grid()
    plt.yticks(fontsize=ytfs)
    plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
    if pnbr in [0, 1, 2]:
        plt.gca().set_xticklabels([])
        if pnbr==2:
            plt.legend(bbox_to_anchor=(1,1), fontsize=15)
            
    elif pnbr in [3, 4, 5]:
        plt.xlabel(xlab)
        if pnbr==3:
            plt.ylabel(r'$|\mathcal{D}_i|/|\mathcal{D}_i|^{(1)}-1$')
```

```python
for p, pd in zip([p, p0], [pd, pd0]):
    Lin = LinData.LinData_Class(p, '/home/robynm/simulations/'+p.sim_name)
    plt.figure(figsize=(20, 10))
    gs = gridspec.GridSpec(2, 3, height_ratios=[2, 1])
    x='an'
    
    plt.subplot(gs[0])
    plt.plot()
    RPlot.plot_locations(x, 'Da', pd)
    RPlot.plot_locations_th(pd[x], pd['t'], Lin.dDa)
    pformat(0)
    
    plt.subplot(gs[1])
    RPlot.plot_locations(x, 'Za', pd)
    RPlot.plot_locations_th(pd[x], pd['t'], Lin.dZa)
    pformat(1)
    
    plt.subplot(gs[2])
    RPlot.plot_locations(x, 'Ca', pd)
    RPlot.plot_locations_th(pd[x], pd['t'], Lin.dCa)
    pformat(2)
    
    plt.subplot(gs[3])
    RPlot.plot_locations(x, 'dDa', pd)
    pformat(3)

    plt.subplot(gs[4])
    RPlot.plot_locations(x, 'dZa', pd)
    pformat(4)

    plt.subplot(gs[5])
    RPlot.plot_locations(x, 'dCa', pd)
    pformat(5)

    plt.subplots_adjust(hspace=0)
    plt.subplots_adjust(wspace=0.25)
```
