```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from data_analysis_codes.tools import Cstyle
plt.style.use(Cstyle.style1)



def det(coord):
    gxx, gxy, gxz, gyy, gyz, gzz = coord
    gdet = -gxz*gxz*gyy + 2*gxy*gxz*gyz - gxx*gyz*gyz - gxy*gxy*gzz + gxx*gyy*gzz
    gup = np.array([[-gyz**2 + gyy*gzz, gxz*gyz - gxy*gzz, -gxz*gyy + gxy*gyz], 
[gxz*gyz - gxy*gzz, -gxz**2 + gxx*gzz, gxy*gxz - gxx*gyz],
[-gxz*gyy + gxy*gyz, gxy*gxz - gxx*gyz, -gxy**2 + gxx*gyy]])
    gup /= gdet
    return gdet, gup

def getvars(path):
    Omega_mi = 1
    G = 1
    ti = 1
    kappa = 8*np.pi*G
    #t
    t, metric, gdown = RRead.readasctable_tensor(path+'admbase-metric.average.asc')
    Hiprop = 2 / ( 3 * ti )
    Hprop = Hiprop * ti / t
    #gdet
    gdet, gup = det(metric)
    #a
    a = np.sqrt(gdet)**(1/3)
    #K
    t, curv, kdown = RRead.readasctable_tensor(path+'admbase-curv.average.asc')
    K = np.einsum('ijk, ijk -> k', gup, kdown)
    #rho, press
    rho, worked = RRead.readasctable_scalar(path+'hydrobase-rho.average.asc')
    press, worked = RRead.readasctable_scalar(path+'hydrobase-press.average.asc')
    if worked:
        print("GRHydro")
    else:
        rho, worked = RRead.readasctable_scalar(path+'ct_dust-ct_rho.average.asc')
        press, worked = RRead.readasctable_scalar(path+'ct_dust-ct_prs.average.asc')
        print("CT_Dust")
    #delta
    rhoflrw = 3 * Hprop**2 * Omega_mi / kappa
    delta = rho/rhoflrw - 1
        
        
    Ham = read_val(path+'ml_bssn-ml_ham.average.asc')
    Mom = read_vec(path+'ml_bssn-ml_mom.average.asc')
    nMom = np.sqrt( Mom[0]**2 + Mom[1]**2 + Mom[2]**2 )
    
    return np.array([t, a, gdet, K, rho, delta, press, Ham, nMom])
```

```python
def L1_error(data):
    t = data[0]
    a = data[2]
    aflrw = t**(2/3)
    rho = data[4]
    Hflrw = 2/(3*t)
    rhoflrw = (3*Hflrw**2)/(8*np.pi)
    K = data[3]
    Kflrw = -3*Hflrw
    N = len(a)
    return [np.sum(np.abs(a/aflrw -1))/N, 
            np.sum(np.abs(rho/rhoflrw -1))/N, 
            np.sum(np.abs(K/Kflrw -1))/N]

Na = np.array([4.0, 5.0, 10.0, 20.0])
L1GRHydro = np.array([L1_error(data_GRHydro_4), L1_error(data_GRHydro_5),
                      L1_error(data_GRHydro_10), L1_error(data_GRHydro_20)])
L1CT_Dust = np.array([L1_error(data_CT_Dust_4), L1_error(data_CT_Dust_5),
                      L1_error(data_CT_Dust_10), L1_error(data_CT_Dust_20)])

Nb = np.array([4.0, 5.0, 10.0, 20.0, 50.0])
L1CT_Dust_ICFLRW = np.array([L1_error(data_CT_Dust_ICFLRW_4), L1_error(data_CT_Dust_ICFLRW_5),
                      L1_error(data_CT_Dust_ICFLRW_10), L1_error(data_CT_Dust_ICFLRW_20),
                             L1_error(data_CT_Dust_ICFLRW_50)])

Nc = np.array([10.0, 20.0, 40.0])
L1CT_Dust_ICFLRW_A0 = np.array([L1_error(data_CT_Dust_ICFLRW_A0_10), L1_error(data_CT_Dust_ICFLRW_A0_20),
                             L1_error(data_CT_Dust_ICFLRW_A0_40)])
#L1CT_Dust_ICFLRW_A1e5 = np.array([L1_error(data_CT_Dust_ICFLRW_A1e5_10), L1_error(data_CT_Dust_ICFLRW_A1e5_20),
#                             L1_error(data_CT_Dust_ICFLRW_A1e5_40)])
Nd = np.array([10.0, 20.0])
#L1CT_Dust_ICFLRW_A1e5_t1 = np.array([L1_error(data_CT_Dust_ICFLRW_A1e5_10), L1_error(data_CT_Dust_ICFLRW_A1e5_20)])
L1CT_Dust_ICFLRW_A0_Ham_t1 = np.array([L1_error(data_CT_Dust_ICFLRW_A0_10_Ham), L1_error(data_CT_Dust_ICFLRW_A0_20_Ham)])
```
