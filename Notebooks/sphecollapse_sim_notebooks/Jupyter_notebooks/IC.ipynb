```python
import numpy as np
import matplotlib.pyplot as plt
from data_analysis_codes import param
from data_analysis_codes.tools import ReadingTools
from data_analysis_codes.tools import RicciS
from matplotlib import ticker
IC = param.pflrw_A1e7_rhoEq56()

def metric(gdown):
    gxx = gdown[0, 0]
    gxy = gdown[0, 1]
    gxz = gdown[0, 2]
    gyx = gdown[1, 0]
    gyy = gdown[1, 1]
    gyz = gdown[1, 2]
    gzx = gdown[2, 0]
    gzy = gdown[2, 1]
    gzz = gdown[2, 2]
    
    gdet = -gxz*gyy*gzx + gxy*gyz*gzx + gxz*gyx*gzy - gxx*gyz*gzy - gxy*gyx*gzz + gxx*gyy*gzz
    
    gup = np.array([[(-gyz*gzy + gyy*gzz), (gxz*gzy - gxy*gzz), (-gxz*gyy + gxy*gyz)], 
    [(gyz*gzx - gyx*gzz), (-gxz*gzx + gxx*gzz), (gxz*gyx - gxx*gyz)], 
    [(-gyy*gzx + gyx*gzy), (gxy*gzx - gxx*gzy), (-gxy*gyx + gxx*gyy)]])/gdet
    
    return gdet, gup

def I3Var(var):
    Box1 = np.ones([IC.N,IC.N,IC.N])
    Box0 = np.zeros([IC.N,IC.N,IC.N])
    if type(var) == float:
        M = 1
    else:
        M = np.shape(var)[0]
    if M == IC.Dim:
        return np.array([[var[0], Box0, Box0],
                         [Box0, var[1], Box0],
                         [Box0, Box0, var[2]]])
    elif M == IC.N:
        return np.array([[var, Box0, Box0],
                         [Box0, var, Box0],
                         [Box0, Box0, var]])
    elif M == 1:
        return var*Box1
    else:
        print('idk what you want')
    
def makeplot(var):
    iz = 0
    var_plot = np.flip(var[:,:,iz], 0)
    xmax = len(var_plot)*IC.dx/2
    plt.figure()
    plt.imshow(var_plot, extent=[-xmax, xmax, -xmax, xmax], cmap=plt.cm.get_cmap('Blues', 11))
    plt.colorbar()
    plt.xlabel(r'$x/L$')
    plt.ylabel(r'$y/L$')
    print('for z =', zz[0,0,iz])
    
def f_xd(p, data):
    dataxd = np.zeros([p.N, p.N])
    for j in range(p.N):
        dataxd[j] = data[:, j, j]
            
    xmax = p.N*p.dx_ref/(2*p.L)
    dmax = xmax*np.sqrt(2)
            
    #plt.figure()
    plt.imshow(dataxd, extent=[-xmax, xmax, -dmax, dmax], cmap=plt.cm.get_cmap('inferno'))
    plt.colorbar(format='%.1e', extend='both', ticks=ticker.MaxNLocator(nbins=9))
    plt.xlabel(r'$x/L$')
    plt.ylabel(r'$d/L$')
```

```python
# Initial Values
Hprop = 2/(3*IC.ti)
Hconf = IC.ai*Hprop
f = IC.Omega_mi**(6/11)
wp = f + (3/2)*IC.Omega_mi

iwpH = 1/(wp*(Hconf**2))
iwpHp = f/(wp*Hconf)
iwpHpp = (-3*IC.Omega_mi+f*(f-(3/2)))/wp+3-f
```

```python
# Define Rc

A = np.ones([IC.Dim,IC.N,IC.N,IC.N])*IC.Amp_pert
k = np.ones([IC.Dim,IC.N,IC.N,IC.N])*IC.k_pert

x = np.arange(-IC.L/2,IC.L/2,IC.dx)
xx,yy,zz = np.meshgrid(x,x,x, indexing='ij')
x3d = np.array([xx,yy,zz])

Rcxyz = np.array([A[i]*np.sin(x3d[i]*k[i]) for i in range(IC.Dim)])
Rc = np.sum(Rcxyz,0)
ddRc = -np.sum(Rcxyz*k**2,0)/IC.ai**2
didjRc = I3Var(-Rcxyz*k**2)

chidown1_Rterm = (didjRc - I3Var(ddRc)/3)

R1conf = 4*ddRc
R1 = R1conf #

makeplot(Rc)
```

```python
imid = int(IC.L/(2*IC.dx))
plt.scatter(x, Rc[:,imid,imid])
plt.grid()
plt.ylim(-IC.Amp_pert, IC.Amp_pert)
```

```python
#METRIC

psi1 = iwpH*ddRc/3+Rc
psi1p = iwpHp*ddRc/3

chi1 = -2*Rc*iwpH
ddchi1 = -2*ddRc*iwpH
chi1p = -2*Rc*iwpHp
chi1pp = -2*Rc*iwpHpp
chidown1 = -2*chidown1_Rterm*iwpH
chidown1p = -2*chidown1_Rterm*iwpHp

gdown = (IC.ai**2)*(I3Var(1-2*psi1)+chidown1)  #for 1st order

gdet, gup = metric(gdown)
gmixed = np.einsum('ij...,jk...->ik...', gup, gdown)
gdet_flrw = IC.ai**6
makeplot(gdet/gdet_flrw)
```

```python
D = RicciS.RicciS_Class(IC)
Christ = D.Christoffel_symbol(gdown, gup)
Rs = D.Ricci_scalar(gup, Christ)
```

```python
makeplot(Rs)
makeplot(R1)
```

```python
plt.plot(x, Rs[:,imid,imid], label='Rs')
plt.plot(x, R1[:,imid,imid], label='R1')
plt.legend()
plt.grid()
```

```python
Rerror = np.mean(abs(Rs-R1))
print("Curvature Error {} for {}^3 points with pert amp of 3x{}".format(Rerror, int(IC.L/IC.dx), IC.Amp_pert))
```

```python
#EXTRINSIC CURVATURE

varthetamixed1 = I3Var(-psi1p)+chidown1p/2
Kmixed = -varthetamixed1-I3Var(Hconf)
Ktrace = np.einsum('ii... -> ...', Kmixed)
theta = -Ktrace
Ktrace_flrw = -3*Hconf
makeplot(Ktrace/Ktrace_flrw)
```

```python
#DENSITY
rho_flrw = (3*IC.Omega_mi*(Hconf**2))/(IC.kappa*(IC.ai**2))
delta1 = ddRc*iwpH
rho = I3Var(rho_flrw)*(1+delta1)
makeplot(rho/rho_flrw)

rhothetaterm = np.einsum('ij...,ji...->...', varthetamixed1, varthetamixed1)
delta1_eq16 = ( 9*(psi1p**2) - rhothetaterm - 12*Hconf*psi1p + R1conf) / (6*(Hconf**2)*IC.Omega_mi)
rho_eq16 = I3Var(rho_flrw)*(1+delta1_eq16)
makeplot(rho_eq16/rho_flrw)
```

```python
plt.plot(x, rho[:,imid,imid]/rho_flrw, label='rho')
plt.plot(x, rho_eq16[:,imid,imid]/rho_flrw, label='rho_eq16')
plt.legend()
plt.grid()
```

```python
deltaerror = np.mean(abs(delta1-delta1_eq16))
print("Density Error {} for {}^3 points with pert amp of 3x{}".format(deltaerror, int(IC.L/IC.dx), IC.Amp_pert))
```

```python
#Comoving GI variables
def norm(f):
    return np.sqrt(f[0]**2+f[1]**2+f[2]**2)

Ca = D.GIvars_Ca(gmixed, IC.ai, Rs)
Ca_norm = norm(Ca)
Da = D.GIvars_Da(gmixed, IC.ai, rho)
Da_norm = norm(Da)
Za = D.GIvars_Za(gmixed, IC.ai, theta)
Za_norm = norm(Za)
```

```python
plt.figure(figsize=(15,7))
f_xd(IC, Ca_norm)
plt.title(r'$|\mathcal{C}_a|$')
plt.figure(figsize=(20,10))
f_xd(IC, Da_norm)
plt.title('Da_norm')
plt.figure(figsize=(20,10))
f_xd(IC, Za_norm)
plt.title('Za_norm')
```
