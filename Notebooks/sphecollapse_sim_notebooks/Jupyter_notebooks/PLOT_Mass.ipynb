```python
from data_analysis_codes import param
from data_analysis_codes.tools import PlottingTools as RPlot

p  = param.pflrw_d3e2_L6_t1_N16_zc200()
pd = RPlot.getdata(p)
```

```python
import matplotlib.pyplot as plt
from data_analysis_codes.tools import Cstyle
plt.style.use(Cstyle.style1)
from matplotlib import gridspec
```

```python
speed_of_light = 299792458   # m.s^{-1}
Grav_const     = 6.67408e-11 # m^3.kg^{-1}.s^{-2}
parsec         = 3.0857e16   # m
Megaparsecc    = parsec*1e6  # m
MassSun        = 1.98847e30  # kg
a_today        = 1 + p.zc
Massfac = (Megaparsecc*speed_of_light**2)/(Grav_const*MassSun*a_today**3)
```

```python
Massfac = (Megaparsecc*speed_of_light**2)/(Grav_const*MassSun*a_today**3)
```

```python
x = 'an'
plt.plot(pd[x], pd['Mass_Total']*Massfac, color='C4', label='Total : {:.2e}'.format(pd['Mass_Total'][0]*Massfac))
plt.plot(pd[x], pd['Mass_ODcube']*Massfac, color='C0', label='ODcube (x & y & z <0)')
plt.plot(pd[x], pd['Mass_ODzone']*Massfac, color='C9', label='ODzone ('+r'$\delta >0$'+')')
plt.plot(pd[x], pd['Mass_UDzone']*Massfac, color='C8', label='UDzone ('+r'$\delta <0$'+')')
plt.plot(pd[x], pd['Mass_UDcube']*Massfac, color='C2', label='UDcube (x & y & z >0)')
plt.plot(pd[x], pd['Mass_TA']*Massfac, color='C1', label='TA ('+r'$\delta K<-1$'+')')
plt.ylabel('Mass '+r'$[M_{\odot}]$')
plt.grid()
plt.xlabel(r'$a/a_{IN}$')
plt.legend(bbox_to_anchor=(1,1), fontsize=15)
```
