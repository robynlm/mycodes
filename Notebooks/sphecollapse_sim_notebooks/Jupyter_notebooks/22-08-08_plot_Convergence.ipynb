```python
def get_error(f32, f64, f128):
    nf32 = np.array(f32)
    nf64 = np.array(f64)[0::2]
    nf128 = np.array(f128)[0::4]
    c = []
    for i in range(min([len(nf32), len(nf64), len(nf128)])):
        if nf64[i] == nf128[i]:
            c += [64]
        else:
            c += [abs(nf32[i]-nf64[i])/abs(nf64[i]-nf128[i])]
    return np.array(c)
```

```python
N = np.array([16.0, 32.0, 64.0, 128.0])
```

```python
sim = param.pflrw_d3e2_L1821_t1_N32_LCDM()
f32 = pd.read_table('/home/robynm/simulations/'+sim.sim_name+'/h5_data.csv', delimiter=',')

sim = param.pflrw_d3e2_L1821_t1_N64_LCDM()
f64 = pd.read_table('/home/robynm/simulations/'+sim.sim_name+'/h5_data.csv', delimiter=',')

sim = param.pflrw_d3e2_L1821_t1_N128_LCDM()
f128 = pd.read_table('/home/robynm/simulations/'+sim.sim_name+'/h5_data.csv', delimiter=',')
Lin128 = LinData.LinData_Class(sim, '/home/robynm/simulations/'+sim.sim_name+'/')

Ham_convergence = get_error(f32[loc], f64[loc], f128[loc])
```
