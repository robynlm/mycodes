```python
def get_delta(sim, p, it):
    filename = '{}_it_{:06d}.hdf5'.format('/home/robynm/simulations/'+sim.sim_name+'/output-0000/pflrw_L1821/all_iterations/'+sim.sim_name, it)
    f = h5py.File(filename, 'r')
    Lin = LinData.LinData_Class(p)
    TAr = TAradius.TA_Class(p, Lin, ([int(p['Nx']/4)], [int(p['Nx']/4)], [int(p['Nx']/4)]))
    get_var = GVar.Get_var(p, Lin)
    
    delta = get_var.get_the_rho(f, it)['drho']
    gdet = get_var.get_the_curv(f, it)['metric_dic']['gdet']
    
    proper_radius_in_three_directions = []
    delta_in_three_directions = []
    for theta, phi, dxfac in zip([0, np.pi/4, np.pi/4], [0, 0, np.pi/4], [1.0, np.sqrt(2), np.sqrt(3)]):
        coord, radius, indices, radius_from_indices, weighted_dx = TAr.idx_along_r(theta, phi, dxfac)[1]
        
        for i in range(len(radius_from_indices)): #this for loop is to find the correct i
            if radius_from_indices[i]>1.05*dxfac*sim.L/2:
                break
                
        delta_in_three_directions += [np.array([delta[i] for i in indices[:i]])]
        gdet_along_radius_save = np.array([gdet[i] for i in indices[:i]])
        weighted_dx_save = weighted_dx[:i]
        
        proper_radius = [0]
        for j in range(1, i):
            proper_radius += [proper_radius[-1]+(gdet_along_radius_save[j]**(1/6))*weighted_dx_save[j]]
        proper_radius_in_three_directions += [np.array(proper_radius)]
    return proper_radius_in_three_directions, delta_in_three_directions
```

```python
r128, d128 = get_delta(sim128, p128, 0)
r64, d64 = get_delta(sim64, p64, 0)
r32, d32 = get_delta(sim32, p32, 0)
```

```python
def interpolate(f):
    f_save = [f[0]]
    for i in range(len(f)-1):
        f_save += [np.average([f[i], f[i+1]]), f[i+1]]
    return np.array(f_save)

def extrapolate(f):
    return np.append(f, f[-1]+f[-1]-f[-2])

def get_error(f32, f64, f128):
    f32 = interpolate(interpolate(f32))
    f64 = interpolate(f64)
    while len(f32)<len(f128):
        f32 = extrapolate(f32)
    while len(f64)<len(f128):
        f64 = extrapolate(f64)
    c = RRead.safe_division(abs(f32-f64), abs(f64-f128))
    return abs((f64-f128)/(c-1))    
```

```python
e_vert = get_error(d32[0], d64[0], d128[0])
e_edge = get_error(d32[1], d64[1], d128[1])
e_face = get_error(d32[2], d64[2], d128[2])
```

```python
def shift_around_OD(data):
    icut = int(np.shape(data)[0]*3/4)
    databig = np.append(data[icut:,:,:], data[:icut,:,:], axis=0)
    databig = np.append(databig[:,icut:,:], databig[:,:icut,:], axis=1)
    return databig
```
