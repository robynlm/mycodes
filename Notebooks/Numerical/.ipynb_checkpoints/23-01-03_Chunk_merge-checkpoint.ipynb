```python
def make_pairs(data, axis):
    ipairs = []
    indices = list(np.arange(len(data)))
    for i in range(len(data)):
        if i in indices:
            pair = [i]
            Nx, Ny, Nz = np.shape(data[i])
            indices.remove(i)
            for j in indices:
                Nxj, Nyj, Nzj = np.shape(data[j])
                if ((axis==0 and Ny==Nyj and Nz==Nzj) or 
                    (axis==1 and Nx==Nxj and Nz==Nzj) or 
                    (axis==2 and Nx==Nxj and Ny==Nyj)):
                    pair += [j]
                    indices.remove(j)
                    break
            ipairs += [pair]
        if len(ipairs)==int(len(data)/2):
            break
    for p in range(len(ipairs)):
        i, j = ipairs[p]
        if axis==0:
            diff0 = np.mean(abs(data[i][-1,:,:]/data[j][0,:,:] - 1))
            diff1 = np.mean(abs(data[j][-1,:,:]/data[i][0,:,:] - 1))
        elif axis==1:
            diff0 = np.mean(abs(data[i][:,-1,:]/data[j][:,0,:] - 1))
            diff1 = np.mean(abs(data[j][:,-1,:]/data[j][:,0,:] - 1))
        else:
            diff0 = np.mean(abs(data[i][:,:,-1]/data[j][:,:,0] - 1))
            diff1 = np.mean(abs(data[j][:,:,-1]/data[j][:,:,0] - 1))
        print(diff1, diff0)
        if diff1<diff0:
            ipairs[p] = [j, i]
    return ipairs
```
