```python
def L_interp(x, y, xf):
    yf = np.zeros(len(xf))
    k = len(x)
    for i in range(len(xf)):
        for j in range(k):
            top = 1
            bot = 1
            for m in range(k):
                if j!=m:
                    top *= xf[i]-x[m]
                    bot *= x[j]-x[m]
            yf[i] += y[j]*top/bot
    return yf

def dichotomomy(y_wanted, function, lower_bound, upper_bound, tolerance):
    x_low = lower_bound
    x_upp = upper_bound
    x_mid = (x_low + x_upp) / 2
    y_low = function(x_low)
    y_upp = function(x_upp)
    y_mid = function(x_mid)
    while abs(y_wanted / y_mid - 1) > tolerance:
        if y_wanted > y_mid:
            y_low = y_mid
            x_low = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
        else:
            y_upp = y_mid
            x_upp = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
    return x_mid
```

```python
def lin_extrap_TA(f):
    imin = np.argmin(abs(f['K_OD']))
    def find_TA(t):
        y = L_interp(np.array(f['t'][imin-3:imin+3]), 
                      np.array(f['K_OD'][imin-3:imin+3]), np.array([t]))
        return y
    tTA = dichotomomy(1e-15, find_TA, 200.0, 220.0, 1e-2)
    deltaTA = L_interp(np.array(f['t'][imin-3:imin+3]), 
                       np.array(f['drho_OD'][imin-3:imin+3]), np.array([tTA]))
    return Lin.an_initial(tTA), deltaTA
    
# TA time
aTAaIN = np.array([lin_extrap_TA(data)[0] for data in [dL, dLsde5, dLsde10, dLsde15, dLsde20]])
deltaTA = np.array([lin_extrap_TA(data)[1] for data in [dL, dLsde5, dLsde10, dLsde15, dLsde20]])
Lx = np.array([paramdata['Lx'] for paramdata in [pL, pLsde5, pLsde10, pLsde15, pLsde20]])
```

```python
plt.figure(figsize=(10,5))

ax1 = plt.subplot(121)
ax2 = plt.subplot(122)
t = np.arange(200, 220, 0.5)
for i, data in enumerate([dL, dLsde10, dLsde15, dLsde20]):
    imin = np.argmin(abs(data['K_OD']))
    ax1.scatter(Lin.an_initial(data['t']), data['K_OD'])
    ax1.plot(Lin.an_initial(t), 
             L_interp(np.array(data['t'][imin-3:imin+3]), 
                      np.array(data['K_OD'][imin-3:imin+3]), t))
    def find_TA(t):
        y = L_interp(np.array(data['t'][imin-3:imin+3]), 
                      np.array(data['K_OD'][imin-3:imin+3]), np.array([t]))
        return y
    tTA = dichotomomy(1e-15, find_TA, 200.0, 220.0, 1e-2)
    ax1.axvline(x=Lin.an_initial(tTA), color='C'+str(i))
    
    ax2.scatter(data['t'], data['drho_OD'])
    ax2.plot(t, 
             L_interp(np.array(data['t'][imin-3:imin+3]), 
                      np.array(data['drho_OD'][imin-3:imin+3]), t))
    deltaTA = L_interp(np.array(data['t'][imin-3:imin+3]), 
                      np.array(data['drho_OD'][imin-3:imin+3]), np.array([tTA]))
    ax2.axhline(y=deltaTA, color='C'+str(i))
    ax2.axvline(x=tTA, color='C'+str(i))
    
ax1.grid()
ax1.set_ylim(-0.0002, 0.0002)
ax1.set_xlim(35.1, 35.5)
ax2.grid()
ax2.set_xlim(208.5, 211)
ax2.set_ylim(4.56, 4.564)
```
