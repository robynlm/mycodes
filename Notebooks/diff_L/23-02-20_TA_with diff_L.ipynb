```python
def L_interp(x, y, xf):
    yf = np.zeros(len(xf))
    k = len(x)
    for i in range(len(xf)):
        for j in range(k):
            top = 1
            bot = 1
            for m in range(k):
                if j!=m:
                    top *= xf[i]-x[m]
                    bot *= x[j]-x[m]
            yf[i] += y[j]*top/bot
    return yf

def dichotomomy(y_wanted, function, lower_bound, upper_bound, tolerance):
    x_low = lower_bound
    x_upp = upper_bound
    x_mid = (x_low + x_upp) / 2
    y_low = function(x_low)
    y_upp = function(x_upp)
    y_mid = function(x_mid)
    while abs(y_wanted / y_mid - 1) > tolerance:
        if y_wanted > y_mid:
            y_low = y_mid
            x_low = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
        else:
            y_upp = y_mid
            x_upp = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
    return x_mid

def lin_extrap_TA(f, Lin, tTAmin):
    imin = np.argmin(abs(f['K_OD']))
    def find_TA(t):
        y = L_interp(np.array(f['t'][imin-3:imin+3]), 
                      np.array(f['K_OD'][imin-3:imin+3]), np.array([t]))
        return y
    tTA = dichotomomy(1e-15, find_TA, tTAmin, tTAmin+20.0, 1e-2)
    try:
        drhoOD = f['drho_OD']
    except:
        drhoOD = f['drho_u_OD']
    deltaTA = L_interp(np.array(f['t'][imin-3:imin+3]), 
                       np.array(drhoOD[imin-3:imin+3]), np.array([tTA]))
    aTAaIN = Lin.evo.a(tTA)/Lin.evo.a(f['t'][0])
    delta1TA = aTAaIN*drhoOD[0]
    return delta1TA, deltaTA[0], aTAaIN

def interpolate(f):
    f_save = [f[0]]
    for i in range(len(f)-1):
        f_save += [np.average([f[i], f[i+1]]), f[i+1]]
    return np.array(f_save)

def extrapolate(f):
    return np.append(f, f[-1]+f[-1]-f[-2])

def get_error(f32, f64, f128):
    if len(f32)!=len(f128):
        f32 = interpolate(interpolate(f32))
        while len(f32)<len(f128):
            f32 = extrapolate(f32)
    if len(f64)!=len(f128):
        f64 = interpolate(f64)
        while len(f64)<len(f128):
            f64 = extrapolate(f64)
    c = abs(f32-f64)/abs(f64-f128)
    err = abs((f64-f128)/(c-1))
    for ic in range(len(c)):
        if err[ic]>abs(f32[ic]-f64[ic]) or err[ic]>abs(f64[ic]-f128[ic]):
            err[ic] = np.max([abs(f32[ic]-f64[ic]), abs(f64[ic]-f128[ic])])
    return err

def get_collapse_delta1(param, Lin, f):
    try:
        tC = np.array(pd.read_table(param['datapath']+'Time_dt.csv', delimiter=',')['t'])[-1]
    except:
        tC = np.array(pd.read_table(param['datapath']+'Time_da.csv', delimiter=',')['t'])[-1]
    try:
        drhoOD = f['drho_OD']
    except:
        drhoOD = f['drho_u_OD']
    aCaIN = Lin.evo.a(tC)/Lin.evo.a(f['t'][0])
    delta1C = aCaIN * drhoOD[0]
    return delta1C
```

```python
d1c = np.array([get_collapse_delta1(paramdata, Lin128, data) for paramdata, data in zip(allparam, alldata)])
d1c64 = np.array([get_collapse_delta1(paramdata, Lin64, data) for paramdata, data in zip(allparam64, alldata64)])
d1c32 = np.array([get_collapse_delta1(paramdata, Lin32, data) for paramdata, data in zip(allparam32, alldata32)])
d1cerr = get_error(d1c32, d1c64, d1c)

d1cEdS = np.array([[get_collapse_delta1(paramdata, LinEdS, data)] for paramdata, data in zip(pEdS, dEdS)])
d1cerrEdS = get_error(d1cEdS[2], d1cEdS[1], d1cEdS[0])

d1cEdS2 = np.array([[get_collapse_delta1(paramdata, LinEdS2, data)] for paramdata, data in zip(pEdS2, dEdS2)])
d1cerrEdS2 = get_error(d1cEdS2[2], d1cEdS2[1], d1cEdS2[0])
```
