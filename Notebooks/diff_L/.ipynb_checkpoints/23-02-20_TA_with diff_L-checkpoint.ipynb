```python
pEdS = [RRead.read_parameters('pflrw_d3e2_L1206_t1_N128_EdS')]
pEdS += [RRead.read_parameters('pflrw_d3e2_L1206_t1_N64_EdS')]
pEdS += [RRead.read_parameters('pflrw_d3e2_L1206_t1_N32_EdS')]
dEdS = [pd.read_table(param['datapath']+'h5_data.csv', delimiter=',') for param in pEdS]
LinEdS = LinData.LinData_Class(pEdS[0])
```

```python
def L_interp(x, y, xf):
    yf = np.zeros(len(xf))
    k = len(x)
    for i in range(len(xf)):
        for j in range(k):
            top = 1
            bot = 1
            for m in range(k):
                if j!=m:
                    top *= xf[i]-x[m]
                    bot *= x[j]-x[m]
            yf[i] += y[j]*top/bot
    return yf

def dichotomomy(y_wanted, function, lower_bound, upper_bound, tolerance):
    x_low = lower_bound
    x_upp = upper_bound
    x_mid = (x_low + x_upp) / 2
    y_low = function(x_low)
    y_upp = function(x_upp)
    y_mid = function(x_mid)
    while abs(y_wanted / y_mid - 1) > tolerance:
        if y_wanted > y_mid:
            y_low = y_mid
            x_low = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
        else:
            y_upp = y_mid
            x_upp = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
    return x_mid

def lin_extrap_TA(f, Lin):
    imin = np.argmin(abs(f['K_OD']))
    def find_TA(t):
        y = L_interp(np.array(f['t'][imin-3:imin+3]), 
                      np.array(f['K_OD'][imin-3:imin+3]), np.array([t]))
        return y
    tTA = dichotomomy(1e-15, find_TA, 200.0, 220.0, 1e-2)
    deltaTA = L_interp(np.array(f['t'][imin-3:imin+3]), 
                       np.array(f['drho_OD'][imin-3:imin+3]), np.array([tTA]))
    aTAaIN = Lin.evo.a(tTA)/Lin.evo.a(f['t'][0])
    delta1TA = aTAaIN*f['drho_OD'][0]
    return delta1TA, deltaTA[0]

def interpolate(f):
    f_save = [f[0]]
    for i in range(len(f)-1):
        f_save += [np.average([f[i], f[i+1]]), f[i+1]]
    return np.array(f_save)

def extrapolate(f):
    return np.append(f, f[-1]+f[-1]-f[-2])

def get_error(f32, f64, f128):
    if len(f32)!=len(f128):
        f32 = interpolate(interpolate(f32))
        while len(f32)<len(f128):
            f32 = extrapolate(f32)
    if len(f64)!=len(f128):
        f64 = interpolate(f64)
        while len(f64)<len(f128):
            f64 = extrapolate(f64)
    c = abs(f32-f64)/abs(f64-f128)
    err = abs((f64-f128)/(c-1))
    for ic in range(len(c)):
        if err[ic]>abs(f32[ic]-f64[ic]) or err[ic]>abs(f64[ic]-f128[ic]):
            err[ic] = np.max([abs(f32[ic]-f64[ic]), abs(f64[ic]-f128[ic])])
    return err
```

```python
Lx = np.array([paramdata['Lx'] for paramdata in allparam])
d1 = np.array([lin_extrap_TA(data, Lin128)[0] for data in alldata])
d164 = np.array([lin_extrap_TA(data, Lin64)[0] for data in alldata64])
d132 = np.array([lin_extrap_TA(data, Lin32)[0] for data in alldata32])
d1err = get_error(d132, d164, d1)

d = np.array([lin_extrap_TA(data, Lin128)[1] for data in alldata])
d64 = np.array([lin_extrap_TA(data, Lin128)[1] for data in alldata64])
d32 = np.array([lin_extrap_TA(data, Lin128)[1] for data in alldata32])
derr = get_error(d32, d64, d)

LxEdS = pEdS[0]['Lx']
d1EdS = np.array([[lin_extrap_TA(data, LinEdS)[0]] for data in dEdS])
d1errEdS = get_error(d1EdS[2], d1EdS[1], d1EdS[0])
deltaEdS = np.array([[lin_extrap_TA(data, LinEdS)[1]] for data in dEdS])
derrEdS = get_error(deltaEdS[2], deltaEdS[1], deltaEdS[0])
```

```python
THdelta1TA = (3/20)*(6*np.pi)**(2/3)
colors = plt.cm.viridis(np.linspace(0, 1, len(Lx)))

plt.plot(Lx, d1, linestyle=':', color='gray', linewidth = 1)
plt.axhline(y = THdelta1TA, color='k')
plt.text(Lx[0]*0.92, 1.0002*THdelta1TA, 'top-hat')

plt.scatter(Lx[0], d1[0], c=colors[0], marker='*', s=100)
plt.errorbar(Lx[0], d1[0], yerr=d1err[0], fmt="b", ecolor='k')

plt.scatter(Lx[1:], d1[1:], c=colors[1:])
plt.errorbar(Lx[1:], d1[1:], yerr=d1err[1:], fmt="b", ecolor='k', linestyle="None")

LxEdS = pEdS[0]['Lx']
plt.scatter(LxEdS, d1EdS[0], c='m', marker='*', s=100)
plt.errorbar(LxEdS, d1EdS[0], yerr=d1errEdS, fmt="b", ecolor='k')

plt.grid()
plt.ylabel(r'$\delta_{OD}^{(1)}$')
plt.xlabel(r'$\lambda_{pert, \; x}$')
```

```python
THdeltaTA = ((3*np.pi/4)**2)-1
colors = plt.cm.viridis(np.linspace(0, 1, len(Lx)))

plt.plot(Lx, d, linestyle=':', color='gray', linewidth = 1)
plt.axhline(y = THdeltaTA, color='k')
plt.text(Lx[0]*0.92, 1.0001*THdeltaTA, 'top-hat')

plt.scatter(Lx[0], d[0], c=colors[0], marker='*', s=100)
plt.errorbar(Lx[0], d[0], yerr=derr[0], fmt="b", ecolor='k')

plt.scatter(Lx[1:], d[1:], c=colors[1:])
plt.errorbar(Lx[1:], d[1:], yerr=derr[1:], fmt="b", ecolor='k', linestyle="None")

plt.scatter(LxEdS, deltaEdS[0], c='m', marker='*', s=100)
plt.errorbar(LxEdS, deltaEdS[0], yerr=derrEdS, fmt="b", ecolor='k')

plt.grid()
plt.ylabel(r'$\delta_{OD}$')
plt.xlabel(r'$\lambda_{pert, \; x}$')
```
