```python
def calc_nonlin(tIN, L, Amp):
    delta, delta1, TotMass, MassOD, gdet, dgdet, RicciS, K, dK, dgdet1, RicciS1, dK1, Mom_stat = A.f_Amp(tIN, [L]*3, [Amp]*3)
    deltanonlin = abs(delta/delta1 - 1)
    dgdetnonlin = abs(dgdet/dgdet1 - 1)
    RicciSnonlin = abs(RicciS/RicciS1 - 1)
    dKnonlin = abs(dK/dK1 - 1)
    return deltanonlin, dgdetnonlin, RicciSnonlin, dKnonlin, MassOD, Mom_stat
```

```python
def power(a):
    if np.sign(np.log(a)/np.log(10))>0:
        return math.ceil(np.log(a)/np.log(10))
    else:
        return math.floor(np.log(a)/np.log(10))
```

```python
def tfroma(a):
    A = 2/(3 *LCDM.Hprop_today*np.sqrt(LCDM.Omega_l_today))
    B = np.sqrt((LCDM.Omega_m_today/LCDM.Omega_l_today)*(a/LCDM.a_today)**(-3)+1)
    tloc= A*np.arctanh(1/B)
    return tloc
def afromz(z):
    return LCDM.a_today/(1+z)
def tfromz(z):
    return tfroma(afromz(z))
```

```python
tINs = np.linspace(tIN_IC, tF, 100)
vals = np.array([calc_nonlin(tIN, L_IC, Amp_IC) for tIN in tINs]).T
zINs = LCDM.z(tINs)
anINs = LCDM.a(tINs)/LCDM.a(tIN_IC)
```

```python
def dichotonomy(y_wanted, function, lower_bound, upper_bound, tolerance):
    x_low = lower_bound
    x_upp = upper_bound
    x_mid = (x_low + x_upp) / 2
    y_low = function(x_low)
    y_upp = function(x_upp)
    y_mid = function(x_mid)
    while abs(y_wanted / y_mid - 1) > tolerance:
        if y_wanted > y_mid:
            y_low = y_mid
            x_low = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
        else:
            y_upp = y_mid
            x_upp = x_mid
            x_mid = (x_low + x_upp) / 2
            y_mid = function(x_mid)
    return x_mid
```

```python
L = 1000
tIN = 1

def get_delta(Amp):
    output = A.f_Amp(tIN, [L]*3, [Amp]*3)
    delta, delta1, TotMass, MassOD, gdet, dgdet, RicciS, K, dK, dgdet1, RicciS1, dK1, Mom_stat = output
    return delta

Amp = dichotonomy(0.03, get_delta, 3e-3, 4e-3, 1e-10)
print(Amp, get_delta(Amp))
```
