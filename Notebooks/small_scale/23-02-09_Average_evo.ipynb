```python
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from data_analysis_codes.tools import LinData
from data_analysis_codes.tools import ReadingTools as RRead
from data_analysis_codes.tools import Cstyle
plt.style.use(Cstyle.style1)
```

```python
flab = ['full sim']
paramA = [RRead.read_parameters('pflrw_d3e2_L1821_t1_N32_LCDM')]
paramA += [RRead.read_parameters('pflrw_d3e2_L1821_t1_N64_LCDM')]
paramA += [RRead.read_parameters('pflrw_d3e2_L1821_t1_N128_LCDM')]
fA = [pd.read_table(p['datapath']+'h5_data.csv', delimiter=',') for p in paramA]

flab += [r'Fix IC with ${}^{(3)}R$ and then $K$']
paramAk = [RRead.read_parameters('pflrw_A79e4_A1e2_L1821_t200_N32_LCDM_moredata')]
paramAk += [RRead.read_parameters('pflrw_A79e4_A1e2_L1821_t200_N64_LCDM_moredata')]
paramAk += [RRead.read_parameters('pflrw_A79e4_A1e2_L1821_t200_N128_LCDM_moredata')]
fAk = [pd.read_table(p['datapath']+'h5_data.csv', delimiter=',') for p in paramAk]

Lin = LinData.LinData_Class(paramA[0])
L = paramA[0]['Lx']
```

```python
def interpolate(f):
    f_save = [f[0]]
    for i in range(len(f)-1):
        f_save += [np.average([f[i], f[i+1]]), f[i+1]]
    return np.array(f_save)

def extrapolate(f):
    return np.append(f, f[-1]+f[-1]-f[-2])

def get_error(f32, f64, f128):
    if len(f32)!=len(f128):
        f32 = interpolate(interpolate(f32))
        while len(f32)<len(f128):
            f32 = extrapolate(f32)
        while len(f32)>len(f128):
            f32 = f32[:-1]
    if len(f64)!=len(f128):
        f64 = interpolate(f64)
        while len(f64)<len(f128):
            f64 = extrapolate(f64)
        while len(f64)>len(f128):
            f64 = f64[:-1]
    c = abs(f32-f64)/abs(f64-f128)
    err = abs((f64-f128)/(c-1))
    for ic in range(len(c)):
        if err[ic]>abs(f32[ic]-f64[ic]) or err[ic]>abs(f64[ic]-f128[ic]):
            err[ic] = np.max([abs(f32[ic]-f64[ic]), abs(f64[ic]-f128[ic])])
    return err
```

```python
allfs = [fA, fAk]
flab = ['full sim']
flab += [r'Fix IC with ${}^{(3)}R$ and then $K$']
loc = 'av'
alphafac = 1

plt.figure(figsize=(25, 7))
plt.subplot(141)
for f in allfs:
    var = 'drho'
    f32 = np.array(f[0][var+'_'+loc])
    f64 = np.array(f[1][var+'_'+loc])
    f128 = np.array(f[2][var+'_'+loc])
    an128 = Lin.evo.a(f[2]['t'])/Lin.evo.a(1)
    plt.plot(an128, f128)
    err = get_error(f32, f64, f128)
    plt.fill_between(an128, f128-err, f128+err, alpha=alphafac)
plt.grid()
plt.title(r'$\langle\delta\rangle_{\mathcal{D}}$')
plt.xlabel(r'$a/a_{IN}$')
#plt.ylim(-1, 30)
plt.xlim(30, 60)

plt.subplot(142)
for f in allfs:
    var = 'K'
    f32 = np.array(f[0][var+'_'+loc])
    f64 = np.array(f[1][var+'_'+loc])
    f128 = np.array(f[2][var+'_'+loc])
    an128 = Lin.evo.a(f[2]['t'])/Lin.evo.a(1)
    plt.plot(an128, f128)
    err = get_error(f32, f64, f128)
    plt.fill_between(an128, f128-err, f128+err, alpha=alphafac)
plt.grid()
plt.title(r'$\langle K\rangle_{\mathcal{D}}$')
plt.xlabel(r'$a/a_{IN}$')
plt.xlim(30, 60)
plt.ylim(-0.012, -0.004)

plt.subplot(143)
for f in allfs:
    var = 'gdet'
    f32 = np.array(f[0][var+'_'+loc])
    f64 = np.array(f[1][var+'_'+loc])
    f128 = np.array(f[2][var+'_'+loc])
    an128 = Lin.evo.a(f[2]['t'])/Lin.evo.a(1)
    plt.plot(an128, f128)
    err = get_error(f32, f64, f128)
    plt.fill_between(an128, f128-err, f128+err, alpha=alphafac)
plt.grid()
plt.title(r'$\langle\gamma\rangle_{\mathcal{D}}$')
plt.xlabel(r'$a/a_{IN}$')
plt.xlim(30, 60)
#plt.ylim(1e-6, 2e-4)

plt.subplot(144)
for f, lab in zip(allfs, flab):
    var = 'RicciS'
    f32 = (Lin.evo.a(f[0]['t'])**2)*np.array(f[0][var+'_'+loc])*(L**2)
    f64 = (Lin.evo.a(f[1]['t'])**2)*np.array(f[1][var+'_'+loc])*(L**2)
    f128 = (Lin.evo.a(f[2]['t'])**2)*np.array(f[2][var+'_'+loc])*(L**2)
    an128 = Lin.evo.a(f[2]['t'])/Lin.evo.a(1)
    plt.plot(an128, f128)
    err = get_error(f32, f64, f128)
    plt.fill_between(an128, f128-err, f128+err, alpha=alphafac)
plt.grid()
plt.title(r'$a^{2} \langle{}^{(3)}R\rangle_{\mathcal{D}} \;\; [\lambda_{pert}^{-2}]$')
plt.xlabel(r'$a/a_{IN}$')
plt.xlim(30, 60)
#plt.ylim(1, 2000)
plt.legend(bbox_to_anchor=(1,1))

plt.subplots_adjust(wspace=0.25)
```
